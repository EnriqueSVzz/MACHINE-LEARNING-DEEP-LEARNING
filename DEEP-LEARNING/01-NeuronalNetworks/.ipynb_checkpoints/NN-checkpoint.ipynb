# Redes Neuronales

### IA vs ML vs Deep Learning

* `IA (Inteligencia Artificial)` es el campo amplio que busca que las mÃĄquinas realicen tareas que requieren inteligencia humana (razonamiento, percepciÃģn, toma de decisionesâĶ).
* `Machine Learning (ML)` es una `subÃĄrea` de la IA que se enfoca en que los `sistemas aprendan de los datos sin ser programados explÃ­citamente`.
* `Deep Learning (DL)` es una `subÃĄrea del ML` que usa `redes neuronales artificiales con muchas capas (âprofundasâ) para aprender` representaciones mÃĄs complejas de los datos.

* Neuronal Networks:
  - `Estructura`: las redes neuronales estÃĄn compuestas por capas de neuronas conectadas, mientras que otros modelos (como ÃĄrboles o regresiones) tienen formas mÃĄs simples o especÃ­ficas.
  - `Aprendizaje mediante pesos`: cada conexiÃģn tiene un peso que se ajusta durante el entrenamiento, y eso es lo que les permite aprender representaciones complejas.

* En `Machine Learning clÃĄsico`, `el modelo suele âextraerâ caracterÃ­sticas de forma explÃ­cita` (por ejemplo, tÃš eliges quÃĐ variables son importantes).
* En `Deep Learning`, la red aprende automÃĄticamente las representaciones relevantes de los datos, `capa por capa`.

* En Deep Learning, tanto las `NN (Neural Networks)` como las `CNN (Convolutional Neural Networks)` son tipos de redes neuronales, pero se diferencian principalmente en cÃģmo procesan los datos y para quÃĐ tipo de problemas son mÃĄs adecuadas.

| CaracterÃ­stica                    | **NN (Neural Network)**                                                                                                       | **CNN (Convolutional Neural Network)**                                                                                               |
| --------------------------------- | ----------------------------------------------------------------------------------------------------------------------------- | ------------------------------------------------------------------------------------------------------------------------------------ |

Las capas convolucionales detectan *patrones espaciales locales*. |

***NOTAS***
* `IA`: El algoritmo de bÃšsqueda de rutas muestra razonamiento automatizado (aunque no necesariamente âaprendeâ).
* `ML`: El modelo de predicciÃģn de ventas usa datos para aprender patrones â aquÃ­ ya aparece el aprendizaje automÃĄtico.
* `DL`: Las redes neuronales para identificar animales aplican aprendizaje profundo sobre imÃĄgenes (un tipo de dato muy complejo).
* Una NN es mÃĄs genÃĐrica: buena para datos sin estructura espacial.
* Una CNN estÃĄ especializada en procesar datos con estructura espacial (como imÃĄgenes), aprovechando la correlaciÃģn entre pÃ­xeles cercanos.

### Modelo Matematico de una NN

* Al recibir una inspiraciÃģn de las neuronas biologicas, debemos dar una entrada a como funcionan estas.
* Como podemos ver en diagrama, la neurona tiene una terminales de recepciÃģn llamadas `Dendritas` y sus correspondientes terminales de salida llamadas `Axones`.
* Las `sinapsis` son las conexiones o puntos de comunicaciÃģn entre dos neuronas. Son fundamentales para el funcionamiento del cerebro y las redes neuronales artificiales (por analogÃ­a).
* Cuando una neurona se `activa`, envÃ­a un impulso elÃĐctrico por su axÃģn que llega hasta las terminales sinÃĄpticas.

| **Componente**            | **Neurona biolÃģgica**                                                                      | **Neurona artificial**                                                                                 | **AnalogÃ­a / FunciÃģn equivalente**                    |
| ------------------------- | ------------------------------------------------------------------------------------------ | ------------------------------------------------------------------------------------------------------ | ----------------------------------------------------- |
| **Dendritas**             | Ramificaciones que **reciben seÃąales** de otras neuronas.                                  | **Entradas (inputs)**: valores numÃĐricos que llegan desde otras neuronas o desde los datos de entrada. | Reciben informaciÃģn (como los sensores o datos).      |
| **Sinapsis**              | Conexiones entre neuronas; transmiten impulsos quÃ­micos.                                   | **Pesos (weights)** que multiplican las entradas.                                                      | Indican la **fuerza o importancia** de cada conexiÃģn. |
| **Soma (cuerpo celular)** | Integra todas las seÃąales recibidas y decide si envÃ­a una seÃąal.                           | **FunciÃģn de suma y activaciÃģn**: combina los valores de entrada y calcula una salida.                 | Decide si âse activaâ o no.                           |
| **AxÃģn**                  | Canal que **envÃ­a la seÃąal elÃĐctrica** a otras neuronas.                                   | **Salida de la neurona** (output).                                                                     | Transmite la informaciÃģn a la siguiente capa.         |
| **Neurotransmisores**     | Sustancias quÃ­micas que modulan la seÃąal enviada.                                          | **FunciÃģn de activaciÃģn** (sigmoid, ReLU, etc.).                                                       | Controlan **cÃģmo** se transmite la seÃąal.             |
| **Plasticidad sinÃĄptica** | Capacidad de la sinapsis para fortalecerse o debilitarse con la experiencia (aprendizaje). | **Ajuste de los pesos** durante el entrenamiento (backpropagation).                                    | Es el **aprendizaje** en la red neuronal.             |

* Al igual que el `perceptrÃģn`, necesitamos la funciÃģn de activaciÃģn
$$
y = f\left( \sum_{i=1}^{n} w_i x_i + b \right)
$$

donde:

$$x_i \text{ son las entradas (inputs)}$$
$$w_i \text{  son los pesos asignados a cada entrada,}$$
$$b \text{  es el sesgo o bias (valor independiente)}$$
$$f(\cdot) \text{ es la funciÃģn de activaciÃģn, tÃ­picamente:}$$

$$
f(z) =
\begin{cases}
1 & \text{si } z \geq 0 \\
0 & \text{si } z < 0
\end{cases}
$$

* ***ÂŋQuÃĐ son las funciones de activaciÃģn Sigmoid, ReLU y Tanh?***
  - Las funciones de activaciÃģn son una parte fundamental de las redes neuronales.
  - Sirven para introducir no linealidad en el modelo, permitiendo que la red aprenda relaciones complejas (no solo combinaciones lineales de las entradas).

* ***SIGMOID*** $$\frac{1}{1 + e^{-x}}$$

* ***TANH*** $$\frac{e^x - e^{-x}}{e^x + e^{-x}}$$

* ***RELU*** $$\max(0, x)$$


| FunciÃģn     | Rango   | Ventajas                                | Desventajas                         | Uso tÃ­pico             |
| ----------- | ------- | --------------------------------------- | ----------------------------------- | ---------------------- |
| **Sigmoid** | (0, 1)  | InterpretaciÃģn probabilÃ­stica           | Gradiente pequeÃąo, no centrada en 0 | Salida binaria         |
| **Tanh**    | (-1, 1) | Centrada en 0, buena para capas ocultas | Gradiente pequeÃąo                   | Capas ocultas clÃĄsicas |
| **ReLU**    | [0, â)  | Simple, rÃĄpida, evita gradiente pequeÃąo | Neuronas muertas                    | CNN, redes profundas   |

### FunciÃģn SoftMax

* La funciÃģn `SoftMax` convierte esos valores en una `distribuciÃģn de probabilidad`, es decir, una serie de nÃšmeros entre 0 y 1 que suman 1.
$$
\text{SoftMax}(z_i) = \frac{e^{z_i}}{\sum_{j=1}^{K} e^{z_j}}
$$

* Podemos ver como para la clase perro esta no pudo ser clasificada correctamente, entonces ÂŋcÃģmo podemos medir que tan buenos son nuestros Weights y Bias?, para esto estÃĄ nuestra funciÃģn Softmax.

* PrÃĄcticamente consiste en transformar el valor en un porcentaje, para esto debemos calcular su exponencial y dividirlo entre la suma de todas las exponenciales de las otras posibles clasificaciones, como resultado tendremos un `%` que indique que tan cerca estamos o lejos estamos.
* Para mejorar estos valores necesitamos nuestra `Loss Function` para castigar.

### FunciÃģn de Costo (Cross Entropy)

* La `Loss Function` mide cuÃĄnto se equivoca el modelo en sus predicciones. En otras palabras, ***compara las salidas predichas por el modelo con los valores reales (etiquetas verdaderas) y devuelve un nÃšmero que representa el âerrorâ.***
* La FunciÃģn Costo es una medida global del error total del modelo sobre todos los datos de entrenamiento.
En muchos casos, la funciÃģn de costo es la media de las pÃĐrdidas individuales (la loss function aplicada a cada muestra).
* Esto es muy importante para evaluar `W` y `b`, la funciÃģn de perdida nos da un indicador de que tan buenos son nuestors pesos y bias.
* En una red neuronal (NN) multiclase, la `Loss Function` mÃĄs utilizada es la `EntropÃ­a Cruzada CategÃģrica (Categorical Cross-Entropy Loss)`.

$$
L(y, \hat{y}) = - \sum_{i=1}^{C} y_i \ln(\hat{y}_i)
$$

* Recordemos que `y gorro` es el valor predicho y convertido a probabilidad (`FunciÃģn SoftMax`).
$$ \hat{y}_i = \frac{e^{z_i}}{\sum_{j=1}^{C} e^{z_j}} $$

* Como ya se mencionÃģ, las NN siguen un `aprendizaje supervisado`, por lo que debemos evaluar nuestras predicciones con los valores que actualmente registrados.
* AquÃ­ ocurre algo interesante, y es que podemos simplificar mÃĄs las formula para este escenario de multiclase.

$$
L(y, \hat{y}) = - \sum_{i=1}^{C} y_i \ln(\hat{y}_i) = - \ln(\hat{y}_i)
$$

* Sabemos que estos son los resultados de la clase `One-Hot`, por lo que a la ahora de seguir la formula vemos como los otros se cancela al multiplicarse por 0.
$$
y =
\begin{bmatrix}
\begin{bmatrix}1 \\ 0 \\ 0\end{bmatrix} &
\begin{bmatrix}0 \\ 1 \\ 0\end{bmatrix} &
\begin{bmatrix}0 \\ 0 \\ 1\end{bmatrix}
\end{bmatrix}
$$

* Algo de la siguientes maneras

$$
\begin{align*}
\text{Gato} &= -((1 * \ln(0.39)) + (0 * \ln(0.28)) + (0 * \ln(0.33))) = -(1 * \ln(0.39)) = 0.94 \\
\text{Perro} &= -((0 * \ln(0.40)) + (1 * \ln(0.17)) + (0 * \ln(0.43))) = -(1 * \ln(0.17)) = 1.77 \\
\text{Pajaro} &= -((0 * \ln(0.23)) + (0 * \ln(0.36)) + (1 * \ln(0.41))) = -(1 * \ln(0.41)) = 0.89
\end{align*}
$$

* Por ultimo, podemos ver como en la grafica de `y=-ln(x)`, podemos ver como los valores para X estÃĄn en el rango de [0, 1], esto tiene un motivo y es que recordemos que estÃĄmos trabajando con probabilidades.
* EstÃĄ funciÃģn penaliza mucho los valores que se alejen mucho del cero o del uno, dependiendo de la clasificaciÃģn, para el gato diÃģ un valor cercano a 0.94, lo cual en la funciÃģn no  darÃ­a un valor de perdida muy cercano a 0.
* Por otro lado para el perro, vemos que darÃ­a un valor algo alto, ahÃ­ es donde nuestro sistema detectarÃ­a que nuestros `W` y `b` no son del todo optimos. 

![NN_CrossEntropy.png](NN_CrossEntropy.png)

* Finalmente, la `funciÃģn costo` representarÃ­a el promedio de todos los valores de la `Lost Function`.
* La funciÃģn costo generalmente se representa con la letra `J`.

$$
J(\theta) = \frac{1}{m} \sum_{i=1}^{M} L(y_i,\hat{y}_i)
$$

### Algoritmo Vector Gradiente Descendiente

* El `Gradient Descent` (***Descenso por Gradiente***) es un algoritmo de `optimizaciÃģn` usado para encontrar los parÃĄmetros (`pesos`) que `minimizan una funciÃģn de costo`.
* Recordemos que previamente encontramos nuestra `funciÃģn de pÃĐrdida o costo (J)`, es decir un parametro indica que tan bueno son nuestros `W` y `b`, pero ahora debemos minimizar al mÃĄximo este valor.
* ***Encontrar los valores de los parÃĄmetros que hacen que el error del modelo sea lo mÃĄs pequeÃąo posible.***


* ***EJEMPLO***
  * Imagina que la funciÃģn de costo `J(Îļ)` es una montaÃąa, y quieres llegar al punto mÃĄs bajo (mÃ­nimo). `El gradiente es lapendiente de esa montaÃąa`.
  * El descenso por gradiente actÃša asÃ­:
    1. Empieza con un valor inicial de parÃĄmetros ð_0.
    2. Calcula el gradiente (la direcciÃģn de mayor aumento de la funciÃģn).
    3. Se mueve en la direcciÃģn contraria al gradiente.
    4. Repite hasta que el cambio sea muy pequeÃąo.}

* Cada iteraciÃģn actualiza los parÃĄmetros de esta forma
$$\theta := \theta - \alpha \, \nabla J(\theta)$$
$$\theta = \text{parÃĄmetros (pesos)}$$
$$\alpha = \text{learning rate (quÃĐ tan grande es el paso ; hiperparÃĄmetro)}$$
$$\, \nabla J(\theta) = \text{gradiente (derivada de la funciÃģn de costo)}$$

* ***IntuiciÃģn simple***
  * `Si el gradiente es grande â estÃĄs en una zona empinada â das un paso mÃĄs grande.`
  * `Si el gradiente es pequeÃąo â estÃĄs acercÃĄndote al mÃ­nimo â das pasos mÃĄs pequeÃąos.`
  * `Si el learning rate es demasiado grande, puedes âsaltarâ el mÃ­nimo.`
  * `Si el learning rate es muy pequeÃąo, el entrenamiento serÃĄ lento.`

* ***GRADIENTE DESCENDIENTE PARA REGRESIÃN LINEAL***
$$
\theta_j := \theta_j - \alpha \, \frac{1}{m} \sum_{i=1}^{m} 
\left( h_\theta(x^{(i)}) - y^{(i)} \right) x_j^{(i)}
$$

* RepresentaciÃģn grÃĄfica del algoritmo gradiente descendiente. (TazÃģn hasta llegar al fondo)
![GradientDescent.png](GradientDescent.png)

* Practicamente es un proceso iterativo hasta encontrar el valor

### ÂŋPor quÃĐ usamos el gradiente para entrenar Redes Neuronales?

* Realmente las `Neuronal Networks` siguen este diseÃąo, es decir una la parte de la multiplicaciÃģn y otra hace la parte de la suma de nuestro `bias`.
* Podemos por asignar unos cuantos valoresy ver el valor que devolverÃ­a, claramente siguiendo nuestra forma
$$ Y = Wx+b $$

* `Backpropagation` es el algoritmo clave que permite que las redes neuronales aprendan.
* Sirve para `ajustar los pesos` de la red para `reducir el error` entre la salida que genera y la salida correcta.
* La red calcula cÃģmo cada peso afectÃģ el error final, esto se hace con `derivadas` (gradientes) `usando la regla de la cadena`.
$$\frac{\partial z}{\partial z}$$

* `ÂŋSi cambio este peso un poquito, cuÃĄnto cambia el error?`, asÃ­, cada peso sabe:
  - si debe aumentar
  - si debe disminuir
  - y en quÃĐ cantidad

* Una vez que tenemos los gradientes, usamos gradiente descendente, recordemos que sigue esta forma:
$$W := W - \alpha \frac{\partial J}{\partial W}$$
* Esto ajusta los pesos para que la prÃģxima predicciÃģn sea mejor. Veamos el siguiente ejemplo

* Para este ejemplo que `Z` depende de muchas variables, `ÂŋquÃĐ pasarÃ­a si decisimos aumentar algÃšn valor?`.
* Debemos revisar que tanto cambia `Z` respecto a todas sus formas.
* Empecemos por `t`.

$$
\begin{aligned}
t &= w x \\
\frac{\partial t}{\partial w} &= x & \text{Valor: 5} \\
\frac{\partial t}{\partial x} &= w & \text{Valor: -2}
\end{aligned}
$$

* Sigamos por `z`, recordemos que `b` es una constante.
* En el Backpropation, donde en cada parcial una variable es variable y la otra es constante.
* `Backprop es derivaciÃģn implÃ­cita + regla de la cadena`.
$$
\begin{aligned}
z &= t + b \\
\frac{\partial z}{\partial t} &= 1 \\
\frac{\partial z}{\partial b} &= 1
\end{aligned}
$$

* Muy bien, ahora tratemos de saber que tanto cambian `w` y `x` repecto a `Z`.
* Reconocer que `z` depende de `w` a travÃĐs de `t`, pero no existe formula directa para `z(w)`.
* Pero sabemos que `z --> t --> x`. Es una funciÃģn compuesta, asÃ­ que usamos la regla de la cadena.
* Sustituir tus derivadas correctas.
* ***La regla de la cadena NO usa exponentes.***
* Cuando ves `exponentes` es porque estÃĄs `derivando potencias`, `no porque estÃĐs aplicando la regla de la cadena`.
$$
\begin{aligned}
z => t => w \\
\frac{\partial z}{\partial w} &= \frac{\partial z}{\partial t} * \frac{\partial t}{\partial w} \\
\frac{\partial z}{\partial w} &= 1 * x &= 1 * 5 &= 5 \\
z => t => x \\
\frac{\partial z}{\partial x} &= \frac{\partial z}{\partial t} * \frac{\partial t}{\partial x} \\
\frac{\partial z}{\partial x} &= 1 * w &= 1 * -2 &= -2
\end{aligned}
$$

![BachPropagation1.png](BachPropagation1.png)

* `ÂŋQuÃĐ significan realmente todas las derivadas en backpropagation?`
* Cada derivada parcial, "Si aumento w en 1 unidad, la pÃĐrdida aumenta aproximadamente 12."
$$\frac{\partial z}{\partial w} = 12 $$
* Por eso en el descenso de gradiente hacemos:
$$W := W - \alpha \frac{\partial J}{\partial W}$$

* Porque queremos movernos en el sentido en que la pÃĐrdida disminuye, y eso lo indica el signo de la derivada.
* Backpropagation usa TODAS estas relaciones para calcular la derivada de z respecto a w, incluso aunque ðĪ estÃĐ varias capas hacia atrÃĄs. Como muchas neuronas estÃĄn encadenadas, se usan productos de derivadas (regla de la cadena).
