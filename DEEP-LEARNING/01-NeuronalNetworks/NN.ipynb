# Redes Neuronales

### IA vs ML vs Deep Learning

* `IA (Inteligencia Artificial)` es el campo amplio que busca que las máquinas realicen tareas que requieren inteligencia humana (razonamiento, percepción, toma de decisiones…).
* `Machine Learning (ML)` es una `subárea` de la IA que se enfoca en que los `sistemas aprendan de los datos sin ser programados explícitamente`.
* `Deep Learning (DL)` es una `subárea del ML` que usa `redes neuronales artificiales con muchas capas (“profundas”) para aprender` representaciones más complejas de los datos.

* Neuronal Networks:
  - `Estructura`: las redes neuronales están compuestas por capas de neuronas conectadas, mientras que otros modelos (como árboles o regresiones) tienen formas más simples o específicas.
  - `Aprendizaje mediante pesos`: cada conexión tiene un peso que se ajusta durante el entrenamiento, y eso es lo que les permite aprender representaciones complejas.

* En `Machine Learning clásico`, `el modelo suele “extraer” características de forma explícita` (por ejemplo, tú eliges qué variables son importantes).
* En `Deep Learning`, la red aprende automáticamente las representaciones relevantes de los datos, `capa por capa`.

* En Deep Learning, tanto las `NN (Neural Networks)` como las `CNN (Convolutional Neural Networks)` son tipos de redes neuronales, pero se diferencian principalmente en cómo procesan los datos y para qué tipo de problemas son más adecuadas.

| Característica                    | **NN (Neural Network)**                                                                                                       | **CNN (Convolutional Neural Network)**                                                                                               |
| --------------------------------- | ----------------------------------------------------------------------------------------------------------------------------- | ------------------------------------------------------------------------------------------------------------------------------------ |
| **Nombre completo**               | Red Neuronal Artificial (ANN o simplemente NN)                                                                                | Red Neuronal Convolucional (CNN)                                                                                                     |
| **Estructura básica**             | Capas totalmente conectadas (“fully connected”) donde cada neurona está conectada a todas las del nivel anterior y siguiente. | Incluye capas convolucionales, de pooling y totalmente conectadas. Las capas convolucionales detectan *patrones espaciales locales*. |
| **Tipo de datos más común**       | Datos tabulares, series temporales o vectores (por ejemplo, predicción de precios, clasificación de texto simple).            | Datos con estructura espacial, como **imágenes**, **video** o incluso **audio**.                                                     |
| **Extracción de características** | La red debe aprender todas las características directamente de los datos —no hay una noción espacial o local implícita.       | Usa filtros (kernels) que extraen automáticamente características locales (bordes, texturas, formas, etc.) de manera jerárquica.     |
| **Parámetros**                    | Tiene muchos parámetros porque cada conexión es independiente.                                                                | Reutiliza pesos (los filtros se aplican en diferentes regiones), lo que reduce drásticamente los parámetros.                         |
| **Ventaja principal**             | Flexibles y fáciles de aplicar a distintos tipos de entradas.                                                                 | Excelente para reconocer patrones visuales o espaciales con alta precisión.                                                          |
| **Ejemplos de uso**               | Predicción numérica, NLP básico, clasificación de datos tabulares.                                                            | Reconocimiento de imágenes, visión por computadora, detección de objetos, reconocimiento facial.                                     |



***NOTAS***
* `IA`: El algoritmo de búsqueda de rutas muestra razonamiento automatizado (aunque no necesariamente “aprende”).
* `ML`: El modelo de predicción de ventas usa datos para aprender patrones — aquí ya aparece el aprendizaje automático.
* `DL`: Las redes neuronales para identificar animales aplican aprendizaje profundo sobre imágenes (un tipo de dato muy complejo).
* Una NN es más genérica: buena para datos sin estructura espacial.
* Una CNN está especializada en procesar datos con estructura espacial (como imágenes), aprovechando la correlación entre píxeles cercanos.

### Modelo Matematico de una NN

* Al recibir una inspiración de las neuronas biologicas, debemos dar una entrada a como funcionan estas.
* Como podemos ver en diagrama, la neurona tiene una terminales de recepción llamadas `Dendritas` y sus correspondientes terminales de salida llamadas `Axones`.
* Las `sinapsis` son las conexiones o puntos de comunicación entre dos neuronas. Son fundamentales para el funcionamiento del cerebro y las redes neuronales artificiales (por analogía).
* Cuando una neurona se `activa`, envía un impulso eléctrico por su axón que llega hasta las terminales sinápticas.

| **Componente**            | **Neurona biológica**                                                                      | **Neurona artificial**                                                                                 | **Analogía / Función equivalente**                    |
| ------------------------- | ------------------------------------------------------------------------------------------ | ------------------------------------------------------------------------------------------------------ | ----------------------------------------------------- |
| **Dendritas**             | Ramificaciones que **reciben señales** de otras neuronas.                                  | **Entradas (inputs)**: valores numéricos que llegan desde otras neuronas o desde los datos de entrada. | Reciben información (como los sensores o datos).      |
| **Sinapsis**              | Conexiones entre neuronas; transmiten impulsos químicos.                                   | **Pesos (weights)** que multiplican las entradas.                                                      | Indican la **fuerza o importancia** de cada conexión. |
| **Soma (cuerpo celular)** | Integra todas las señales recibidas y decide si envía una señal.                           | **Función de suma y activación**: combina los valores de entrada y calcula una salida.                 | Decide si “se activa” o no.                           |
| **Axón**                  | Canal que **envía la señal eléctrica** a otras neuronas.                                   | **Salida de la neurona** (output).                                                                     | Transmite la información a la siguiente capa.         |
| **Neurotransmisores**     | Sustancias químicas que modulan la señal enviada.                                          | **Función de activación** (sigmoid, ReLU, etc.).                                                       | Controlan **cómo** se transmite la señal.             |
| **Plasticidad sináptica** | Capacidad de la sinapsis para fortalecerse o debilitarse con la experiencia (aprendizaje). | **Ajuste de los pesos** durante el entrenamiento (backpropagation).                                    | Es el **aprendizaje** en la red neuronal.             |

* Al igual que el `perceptrón`, necesitamos la función de activación
$$
y = f\left( \sum_{i=1}^{n} w_i x_i + b \right)
$$

donde:

$$x_i \text{ son las entradas (inputs)}$$
$$w_i \text{  son los pesos asignados a cada entrada,}$$
$$b \text{  es el sesgo o bias (valor independiente)}$$
$$f(\cdot) \text{ es la función de activación, típicamente:}$$

$$
f(z) =
\begin{cases}
1 & \text{si } z \geq 0 \\
0 & \text{si } z < 0
\end{cases}
$$

* ***¿Qué son las funciones de activación Sigmoid, ReLU y Tanh?***
  - Las funciones de activación son una parte fundamental de las redes neuronales.
  - Sirven para introducir no linealidad en el modelo, permitiendo que la red aprenda relaciones complejas (no solo combinaciones lineales de las entradas).

* ***SIGMOID*** $$\frac{1}{1 + e^{-x}}$$

* ***TANH*** $$\frac{e^x - e^{-x}}{e^x + e^{-x}}$$

* ***RELU*** $$\max(0, x)$$


| Función     | Rango   | Ventajas                                | Desventajas                         | Uso típico             |
| ----------- | ------- | --------------------------------------- | ----------------------------------- | ---------------------- |
| **Sigmoid** | (0, 1)  | Interpretación probabilística           | Gradiente pequeño, no centrada en 0 | Salida binaria         |
| **Tanh**    | (-1, 1) | Centrada en 0, buena para capas ocultas | Gradiente pequeño                   | Capas ocultas clásicas |
| **ReLU**    | [0, ∞)  | Simple, rápida, evita gradiente pequeño | Neuronas muertas                    | CNN, redes profundas   |

### Función SoftMax

* La función `SoftMax` convierte esos valores en una `distribución de probabilidad`, es decir, una serie de números entre 0 y 1 que suman 1.
$$
\text{SoftMax}(z_i) = \frac{e^{z_i}}{\sum_{j=1}^{K} e^{z_j}}
$$

* Podemos ver como para la clase perro esta no pudo ser clasificada correctamente, entonces ¿cómo podemos medir que tan buenos son nuestros Weights y Bias?, para esto está nuestra función Softmax.

* Prácticamente consiste en transformar el valor en un porcentaje, para esto debemos calcular su exponencial y dividirlo entre la suma de todas las exponenciales de las otras posibles clasificaciones, como resultado tendremos un `%` que indique que tan cerca estamos o lejos estamos.
* Para mejorar estos valores necesitamos nuestra `Loss Function` para castigar.

### Función de Costo (Cross Entropy)

* La `Loss Function` mide cuánto se equivoca el modelo en sus predicciones. En otras palabras, ***compara las salidas predichas por el modelo con los valores reales (etiquetas verdaderas) y devuelve un número que representa el “error”.***
* La Función Costo es una medida global del error total del modelo sobre todos los datos de entrenamiento.
En muchos casos, la función de costo es la media de las pérdidas individuales (la loss function aplicada a cada muestra).
* Esto es muy importante para evaluar `W` y `b`, la función de perdida nos da un indicador de que tan buenos son nuestors pesos y bias.
* En una red neuronal (NN) multiclase, la `Loss Function` más utilizada es la `Entropía Cruzada Categórica (Categorical Cross-Entropy Loss)`.

$$
L(y, \hat{y}) = - \sum_{i=1}^{C} y_i \ln(\hat{y}_i)
$$

* Recordemos que `y gorro` es el valor predicho y convertido a probabilidad (`Función SoftMax`).
$$ \hat{y}_i = \frac{e^{z_i}}{\sum_{j=1}^{C} e^{z_j}} $$

* Como ya se mencionó, las NN siguen un `aprendizaje supervisado`, por lo que debemos evaluar nuestras predicciones con los valores que actualmente registrados.
* Aquí ocurre algo interesante, y es que podemos simplificar más las formula para este escenario de multiclase.

$$
L(y, \hat{y}) = - \sum_{i=1}^{C} y_i \ln(\hat{y}_i) = - \ln(\hat{y}_i)
$$

* Sabemos que estos son los resultados de la clase `One-Hot`, por lo que a la ahora de seguir la formula vemos como los otros se cancela al multiplicarse por 0.
$$
y =
\begin{bmatrix}
\begin{bmatrix}1 \\ 0 \\ 0\end{bmatrix} &
\begin{bmatrix}0 \\ 1 \\ 0\end{bmatrix} &
\begin{bmatrix}0 \\ 0 \\ 1\end{bmatrix}
\end{bmatrix}
$$

* Algo de la siguientes maneras

$$
\begin{align*}
\text{Gato} &= -((1 * \ln(0.39)) + (0 * \ln(0.28)) + (0 * \ln(0.33))) = -(1 * \ln(0.39)) = 0.94 \\
\text{Perro} &= -((0 * \ln(0.40)) + (1 * \ln(0.17)) + (0 * \ln(0.43))) = -(1 * \ln(0.17)) = 1.77 \\
\text{Pajaro} &= -((0 * \ln(0.23)) + (0 * \ln(0.36)) + (1 * \ln(0.41))) = -(1 * \ln(0.41)) = 0.89
\end{align*}
$$

* Por ultimo, podemos ver como en la grafica de `y=-ln(x)`, podemos ver como los valores para X están en el rango de [0, 1], esto tiene un motivo y es que recordemos que estámos trabajando con probabilidades.
* Está función penaliza mucho los valores que se alejen mucho del cero o del uno, dependiendo de la clasificación, para el gato dió un valor cercano a 0.94, lo cual en la función no  daría un valor de perdida muy cercano a 0.
* Por otro lado para el perro, vemos que daría un valor algo alto, ahí es donde nuestro sistema detectaría que nuestros `W` y `b` no son del todo optimos. 

![NN_CrossEntropy.png](NN_CrossEntropy.png)

* Finalmente, la `función costo` representaría el promedio de todos los valores de la `Lost Function`.
* La función costo generalmente se representa con la letra `J`.

$$
J(\theta) = \frac{1}{m} \sum_{i=1}^{M} L(y_i,\hat{y}_i)
$$

### Algoritmo Vector Gradiente Descendiente

* El `Gradient Descent` (***Descenso por Gradiente***) es un algoritmo de `optimización` usado para encontrar los parámetros (`pesos`) que `minimizan una función de costo`.
* Recordemos que previamente encontramos nuestra `función de pérdida o costo (J)`, es decir un parametro indica que tan bueno son nuestros `W` y `b`, pero ahora debemos minimizar al máximo este valor.
* ***Encontrar los valores de los parámetros que hacen que el error del modelo sea lo más pequeño posible.***


* ***EJEMPLO***
  * Imagina que la función de costo `J(θ)` es una montaña, y quieres llegar al punto más bajo (mínimo). `El gradiente es lapendiente de esa montaña`.
  * El descenso por gradiente actúa así:
    1. Empieza con un valor inicial de parámetros 𝜃_0.
    2. Calcula el gradiente (la dirección de mayor aumento de la función).
    3. Se mueve en la dirección contraria al gradiente.
    4. Repite hasta que el cambio sea muy pequeño.}

* Cada iteración actualiza los parámetros de esta forma
$$\theta := \theta - \alpha \, \nabla J(\theta)$$
$$\theta = \text{parámetros (pesos)}$$
$$\alpha = \text{learning rate (qué tan grande es el paso ; hiperparámetro)}$$
$$\, \nabla J(\theta) = \text{gradiente (derivada de la función de costo)}$$

* ***Intuición simple***
  * `Si el gradiente es grande → estás en una zona empinada → das un paso más grande.`
  * `Si el gradiente es pequeño → estás acercándote al mínimo → das pasos más pequeños.`
  * `Si el learning rate es demasiado grande, puedes “saltar” el mínimo.`
  * `Si el learning rate es muy pequeño, el entrenamiento será lento.`

* ***GRADIENTE DESCENDIENTE PARA REGRESIÓN LINEAL***
$$
\theta_j := \theta_j - \alpha \, \frac{1}{m} \sum_{i=1}^{m} 
\left( h_\theta(x^{(i)}) - y^{(i)} \right) x_j^{(i)}
$$

### ¿Por qué usamos el gradiente para entrenar Redes Neuronales?

* Realmente las `Neuronal Networks` siguen este diseño, es decir una la parte de la multiplicación y otra hace la parte de la suma de nuestro `bias`.
* Podemos por asignar unos cuantos valoresy ver el valor que devolvería, claramente siguiendo nuestra forma
$$ Y = Wx+b $$

* `Backpropagation` es el algoritmo clave que permite que las redes neuronales aprendan.
* Sirve para `ajustar los pesos` de la red para `reducir el error` entre la salida que genera y la salida correcta.
* La red calcula cómo cada peso afectó el error final, esto se hace con `derivadas` (gradientes) `usando la regla de la cadena`.
$$\frac{\partial z}{\partial z}$$

* `¿Si cambio este peso un poquito, cuánto cambia el error?`, así, cada peso sabe:
  - si debe aumentar
  - si debe disminuir
  - y en qué cantidad

* Una vez que tenemos los gradientes, usamos gradiente descendente, recordemos que sigue esta forma:
$$W := W - \alpha \frac{\partial J}{\partial W}$$
* Esto ajusta los pesos para que la próxima predicción sea mejor. Veamos el siguiente ejemplo

* Para este ejemplo que `Z` depende de muchas variables, `¿qué pasaría si decisimos aumentar algún valor?`.
* Debemos revisar que tanto cambia `Z` respecto a todas sus formas.
* Empecemos por `t`.

$$
\begin{aligned}
t &= w x \\
\frac{\partial t}{\partial w} &= x & \text{Valor: 5} \\
\frac{\partial t}{\partial x} &= w & \text{Valor: -2}
\end{aligned}
$$

* Sigamos por `z`, recordemos que `b` es una constante.
* En el Backpropation, donde en cada parcial una variable es variable y la otra es constante.
* `Backprop es derivación implícita + regla de la cadena`.
$$
\begin{aligned}
z &= t + b \\
\frac{\partial z}{\partial t} &= 1 \\
\frac{\partial z}{\partial b} &= 1
\end{aligned}
$$

* Muy bien, ahora tratemos de saber que tanto cambian `w` y `x` repecto a `Z`.
* Reconocer que `z` depende de `w` a través de `t`, pero no existe formula directa para `z(w)`.
* Pero sabemos que `z --> t --> x`. Es una función compuesta, así que usamos la regla de la cadena.
* Sustituir tus derivadas correctas.
* ***La regla de la cadena NO usa exponentes.***
* Cuando ves `exponentes` es porque estás `derivando potencias`, `no porque estés aplicando la regla de la cadena`.
$$
\begin{aligned}
z => t => w \\
\frac{\partial z}{\partial w} &= \frac{\partial z}{\partial t} * \frac{\partial t}{\partial w} \\
\frac{\partial z}{\partial w} &= 1 * x &= 1 * 5 &= 5 \\
z => t => x \\
\frac{\partial z}{\partial x} &= \frac{\partial z}{\partial t} * \frac{\partial t}{\partial x} \\
\frac{\partial z}{\partial x} &= 1 * w &= 1 * -2 &= -2
\end{aligned}
$$

![BachPropagation1.png](BachPropagation1.png)

* `¿Qué significan realmente todas las derivadas en backpropagation?`
* Cada derivada parcial, "Si aumento w en 1 unidad, la pérdida aumenta aproximadamente 12."
$$\frac{\partial z}{\partial w} = 12 $$
* Por eso en el descenso de gradiente hacemos:
$$W := W - \alpha \frac{\partial J}{\partial W}$$

* Porque queremos movernos en el sentido en que la pérdida disminuye, y eso lo indica el signo de la derivada.
* Backpropagation usa TODAS estas relaciones para calcular la derivada de z respecto a w, incluso aunque 𝑤 esté varias capas hacia atrás. Como muchas neuronas están encadenadas, se usan productos de derivadas (regla de la cadena).
