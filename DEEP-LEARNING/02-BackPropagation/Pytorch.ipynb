* `import torch.nn as nn`: Contiene clases de capas y modelos para redes neuronales.
  - `nn.Linear` – capas densas.
  - `nn.Conv2d` – capas convolucionales.
  - `nn.ReLU`, `nn.Sigmoid`, `nn.Softmax` – activaciones (como objetos)
  - `nn.Sequential` – modelos secuenciales.
  - `nn.Module` – clase base para crear tus propios modelos
  - `nn.Dropout`, `nn.BatchNorm2d`, etc.
  - Si haces tu propia red, la clase hereda de `nn.Module`.

* `import torch.nn.functional as F`: Contiene la versión ***"funcional"*** de las funciones de nn.
* Ejemplo:
  - `nn.ReLU()` → capa como objeto
  - `F.relu(x)` → función aplicada directamente
* Incluye:
  - `Activaciones`: F.relu, F.sigmoid, F.softmax
  - `Operaciones de convolución sin crear capas`: F.conv2d
  - `Funciones de pérdida (some)`: F.cross_entropy, F.mse_loss
  - `Pooling`: F.max_pool2d, F.avg_pool2d
  - Se usa cuando quieres más control o no quieres crear una capa como objeto.

* `from torch.utils.data import DataLoader`: Sirve para crear cargadores de datos.
  - Dividir el dataset en lotes (mini-batches)
  - Shuffle (mezclar)
  - Cargar datos en paralelo (num_workers)
  - Iterar con facilidad

* `from torch.utils.data import sampler`: Contiene herramientas para muestrear datos de forma personalizada. Se usa para controlar qué muestras salen y en qué orden.

* `import torchvision.datasets as dataset`: Contiene datasets ya preparados para visión computacional.
  - dataset.MNIST
  - dataset.CIFAR10
  - dataset.ImageFolder
  - dataset.FashionMNIST
  - dataset.CelebA

* `import torchvision.transforms as T`: Contiene transformaciones para imágenes. Sirve para preprocesar los datos antes de alimentar a la red.
  - T.ToTensor() – convierte una imagen a tensor
  - T.Resize()
  - T.Normalize()
  - T.RandomHorizontalFlip()
  - T.Compose([...]) – para agrupar transformaciones

## Compute Accuracy

````python
def compute_acc(loader, model, eval_mode=False):
    num_correct = 0
    num_total = 0
    if eval_mode: model.eval()
    with torch.no_grad():
        for x, y in loader:
            x = x.to(device=device, dtype=dtype)
            y = y.to(device=device, dtype=torch.long)
            scores = model(x)
            _, pred = scores.max(1)
            assert pred.shape == y.shape, 'Label shape and prediction shape does not match'
            num_correct += (pred==y).sum()
            num_total += pred.size(0)           
            
        return float(num_correct)/num_total
````

````python
def train_loop(model, epochs=10, learning_rate = 1e-6):
#     model = model.to(device=device)
    losses = []
    num_batches = len(loader_train)
    for epoch in range(epochs):
        acum_loss = 0.
        for it, (x, y) in enumerate(loader_train):
            #Move to selected device
            x=x.to(device=device, dtype=dtype)
            y=y.to(device=device, dtype=torch.long)
#             print(x.shape)
            #compute scores
#             scores = model.forward(x)
            scores = model(x)
            #compute cost
            cost = F.cross_entropy(input=scores, target=y)
            #compute gradient
            cost.backward()
            
            #update parameters
            with torch.no_grad():
                model.W1 -= learning_rate * model.W1.grad
                model.b1 -= learning_rate * model.b1.grad
                
                model.W2 -= learning_rate * model.W2.grad
                model.b2 -= learning_rate * model.b2.grad
                
                #zero grads
                model.W1.grad.zero_()
                model.b1.grad.zero_()
                model.W2.grad.zero_()
                model.b2.grad.zero_()            
            
            acum_loss += cost.item()
        losses.append(acum_loss/num_batches)
        print(f'Epoch: {epoch}, loss: {cost.item()}, val accuracy: {compute_acc(loader_val, model)} ')
        print()
            
````
