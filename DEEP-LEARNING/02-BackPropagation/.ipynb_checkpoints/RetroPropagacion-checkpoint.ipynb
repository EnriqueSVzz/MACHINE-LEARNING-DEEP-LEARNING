## Recordando...

* Recordemos que el vector gradiente nos permite ir recorriendo para llegar a la pérdida minima, en este caso tendríamos que estar ajustando nuestros parametros `W, b`, estos cambios se dan de la siguiente manera

* Es nuestros `pesos sinapticos(W)` actuales menos el cambio de nuestros pesos en `J` por un `learning Rate`.
$$W = W - \alpha \frac{\partial J}{\partial W}$$

* Nuestro `bias` está dado de la misma manera
$$b = b - \alpha \frac{\partial J}{\partial b}$$

* Recordemos que un valor pequeño de `alpha` puede hacer que tardemos muchisimo, mientra que un valor muy grande haría que demos pasos demasiado grandes.

## Gradiente y el "Cross Entropy"

* La relación entre el `gradiente de la función de pérdida Cross Entropy` y la `función Softmax` `es` uno de los `resultados` más `importantes` y elegantes del `aprendizaje profundo`, porque produce una expresión de `gradiente simple y estable que facilita el entrenamiento de redes neuronales clasificadoras`.
* Recordemos que `Softmax convierte nuestros resultados en probabilidades`.
$$ \hat{y_i} = \text{SoftMax}(z_i) = \frac{e^{z_i}}{\sum_{j=1}^{K} e^{z_j}}$$

* Recordemo que `Cross Entropy` mide la discrepancia entre la predicción y la etiqueta. En otras palabras, ***compara las salidas predichas por el modelo con los valores reales (etiquetas verdaderas) y devuelve un número que representa el “error”.***
* En una red neuronal (NN) multiclase, la `Loss Function` más utilizada es la `Entropía Cruzada Categórica (Categorical Cross-Entropy Loss)`.
* Para un solo ejemplo con etiqueta verdadera 𝑦 (codificada en one-hot), la pérdida es:
$$L(y, \hat{y}) = - \sum_{i=1}^{C} y_i \ln(\hat{y}_i)$$

* Si la clase correcta es 𝑐 entonces:
$$L= - \ln(\hat{y}_c)$$

* La relación fundamental: Gradiente de Cross Entropy + Softmax.
$$
\begin{aligned}
\frac{\partial L}{\partial z_i} &= y - \hat{y_i}
\end{aligned}
$$

* ***¿Qué significa esto?***
  - El gradiente es simplemente la diferencia entre la probabilidad predicha y la etiqueta real.
  - Este resultado sólo es así porque combinamos Softmax + Cross-Entropy.

## Derivada de Gradiente y el "Cross Entropy"

* ***PASO 1*** — Diferenciamos la Cross Entropy respecto `y gorro i`.
* Tomamos solo un término del sumatorio y lo derivamos respecto a `y gorro i`.
$$ Cross Entropy = L(y, \hat{y}) = - \sum_{i=1}^{C} y_i \ln(\hat{y}_i)$$
$$ L_i = - y_i \ln(\hat{y}_i)$$

* Recordemos las derivadas de `logaritmos`.
$$\begin{aligned} \frac{\partial L_i}{\partial\hat{y}_i} &= - y_i \cdot \frac{1}{\hat{y}_i} = - \frac{y_i}{\hat{y}_i} \end{aligned}$$

* ***PASO 2*** — Ahora derivamos el Softmax, entonces estamos buscando la siguiente derivada parcial.
$$\begin{aligned} \frac{\partial \hat{y_i}}{\partial\hat{z}_k} \end{aligned}$$

* Recordemos la definición de la función `SoftMax`, para agilizar un poco definamos una variable `S` la que contendrá el denominador.
$$\hat{y_i} = \frac{e^{z_i}}{\sum_{j=1}^{K} e^{z_j}}$$

$$S = \sum_{j=1}^{K} e^{z_j} $$

$$ \hat{y_i} = \frac{e^{z_i}}{S}$$

* Volvemos a escoger un elemento de la iteración y calculamos que tanto cambia `z_i`, por la forma en la que sigue la derivación necesitaría la `regla del cociente`.

$$\left(\frac{u}{v}\right)'=\frac{v\,u' - u\,v'}{v^{2}}$$

$$
\begin{aligned} \\
\frac{\partial \hat{y_i}}{\partial{z_i}} &= \frac{\partial}{\partial{z_i}} \left(\frac{e^{z_i}}{S}\right)
\end{aligned}
$$

* ***Caso 2: i==k***
* Donde cada valor sería (***NO OLVIDAR QUE 'S' también es una sumaria de 'e'***):
$$
\begin{aligned} \\
u &= e^{z_i}  & u' &= e^{z_i} \\
v &= S & v' &= \frac{\partial{S}}{\partial{z_i}} = e^{z_i} \\
\end{aligned}
$$

* Sustituyamos los valores en la formula y factorizamos:

$$\frac{\partial \hat{y_i}}{\partial{z_i}} = \frac{S e^{z_i} - e^{z_i} e^{z_i} }{S^{2}} = \frac{e^{z_i} \left(S - e^{z_i}\right)}{S^{2}}$$

* Podemos ver si dividimos entre S las 2 fracciones no da la forma de `y gorro i`.
$$ = \left(\frac{e^{z_i}}{S} \right) \cdot \left(1 - \frac{e^{z_i}}{S} \right) = \hat{y_i} \left(1 - \hat{y_i}\right) $$

* ***Caso 2: i!=k***

$$
\begin{aligned} \\
u &= e^{z_i}  & u' &= 0 & \text{Sería la derivada de una constante} \\
v &= S & v' &= e^{z_k} \\
\end{aligned}
$$

* Sustituyamos los valores en la formula y factorizamos:

$$\frac{\partial \hat{y_i}}{\partial{z_k}} = \frac{ S \cdot 0  - e^{z_i} e^{z_k} }{S^{2}} = - \frac{e^{z_i} e^{z_k}}{S^{2}}$$

* Simplificamos diviendo entre S:
$$ = - \frac{e^{z_i}}{S} \cdot \frac{e^{z_k}}{S} = - \hat{y_i} \hat{y_k}$$

## EJEMPLO COMPLETO (Front y Back Propagation)

* Recordemo que habiamos usado este modelo de clasificador multiclase.
* Recordemos que esta parte hace alusión a la primera fase que es la `FrontPropagation`, es decit hacer el calculo de `Z`, transformala a una probabiliad con `softmax` y posterior calcular su `Lost Function`.
* Algo que no noté en el ejemplo pasado, fue que hay 2 `j` y `k`, en la primera representa cualquier iteración donde no es el resultado correcto y en el segundo es la comparación con la clase correcta.
* Esto porque recordemos que `y_target` está códificado como un `OneHotVector`.
* Tomando los valores del ejemplo nuestro primer `Front` quedaría de la siguiente manera.
* Recordemos que si sumamos todos los `softmax` es 1.
* Realmente el que nos importa es el de la clase verdadera, ya que en este ejemplo el algoritmo detectó que era un `ave` y no un `perro`, de hecho su probabilidad fue la más baja.

![Esqueleto.png](Esqueleto.png)
![Front1.png](Front1.png)

* Muy bien en nuestro primer `Front` nos da un valor de `1.0998`, ahora debemos hacer el `BackPropagation`.
* Recordemos que la derivada esta dada por:
$$\begin{aligned} \frac{\partial L}{\partial z} \end{aligned} = \hat{y} - y$$
* Aunque también el algoritmo debe calcular las derivadas respecto `W` y `b` conforme a la `L`.

$$
\begin{aligned} \\
\frac{\partial L}{\partial z} &= \hat{y} - y \\
\\
W &= W - \alpha \cdot \frac{\partial L}{\partial W} \\
\frac{\partial L}{\partial W} &= \frac{\partial L}{\partial z} \cdot \frac{\partial z}{\partial W} \\
\\
b &= b - \alpha \cdot \frac{\partial L}{\partial b} \\
\frac{\partial L}{\partial b} &= \frac{\partial L}{\partial z} \cdot \frac{\partial z}{\partial b} \\
\end{aligned}
$$


![Back1.png](Back1.png)

* Ahora si empecemos por sustituirlo por los valores que tenemos.
* Empecemos por calcular la derivada parcial de `Z` sobre `L`.
* Sería `y = [0.3334, 0.3329, 0.3337] - [0, 1 , 0] = [0.3334, -0.6671, 0.3337]`.
* Tomaremos nuestro `learning rate` como `0.01`.
* Podemos ver como nuestra derivada sería un vector con 3 valores, en este caso 1 por cada clase.
* Ahora encontremos como ajustar el valor `W` y `b`.

$$
\begin{aligned} \\
z&= W \cdot x + b \\
\frac{\partial z}{\partial W} &= X & \frac{\partial z}{\partial b} &= 1 \\
\\
\frac{\partial L}{\partial W} &= \frac{\partial L}{\partial z} \cdot \frac{\partial z}{\partial W} & \frac{\partial L}{\partial b} &= \frac{\partial L}{\partial z} \cdot \frac{\partial z}{\partial b} \\
W &= W - \alpha \cdot \frac{\partial L}{\partial W} \\
b &= b - \alpha \cdot \frac{\partial L}{\partial b} \\
\end{aligned}
$$

* Sustiyendo Valores:
$$
\begin{aligned} \\
W &= W - \alpha \cdot \frac{\partial L}{\partial z} \frac{\partial z}{\partial W}\\
b &= b - \alpha \cdot \frac{\partial L}{\partial z} \frac{\partial z}{\partial b}\\
\end{aligned}
$$

$$
\begin{aligned}
W &= W - \alpha \frac{\partial L}{\partial z} X^T \\
b &= b - \alpha \frac{\partial L}{\partial z}
\end{aligned}
$$

* Realmente podemos `X Transpuesta` por el motivo que para poder hacer la multiplicación deben enbonar, por lo que que si vemos `dL/dZ` tiene una dimensión de `(3,1)` y nuestro input es de `(4,1)`, si hacemos su transpuesta podemos notar como nos daría una matriz de dimensionalidad `(3,4)` que es justo el tamaño que tiene `W`.

$$
\begin{aligned}
W &= W - 0.01 \cdot 
\begin{bmatrix}
0.3334 & -0.6671 & 0.3337
\end{bmatrix}
X^T \\
b &= b - 0.01 \cdot 
\begin{bmatrix}
0.3334 & -0.6671 & 0.3337
\end{bmatrix}
\end{aligned}
$$


```python
import numpy as np

W = np.array([
    [0.0032, -0.0266, 0.0006, -0.0097],
    [0.0036, -0.0101, -0.0081, 0.0084],
    [-0.0006, 0.005, -0.0065, 0.0265]
])

X = np.array([
    [0.1852], 
    [0.0754], 
    [0.437], 
    [0.0619]
])

b = np.array([0., 0., 0.])

dL_dz = np.array([0.3334, -0.6671, 0.3337])

# Gradientes correctos
dL_dW = dL_dz[:, None] @ X.T
dL_db = dL_dz

# Actualización
lr = 0.001
W_new = W - lr * dL_dW
b_new = b - lr * dL_db

print("Nueva W:\n", W_new)
print("Nuevo b:\n", b_new)
```

## Código

```python
import numpy as np

def Softmax(x):
    exp_x = np.exp(x)
    sum_exp = np.sum(exp_x) 
    return exp_x / sum_exp

# Generar nuestros cuatro pixeles sintéticos
# Transpuesta para ponerlo como vector columna
x = np.array([[ 0.1852, 0.0754, 0.437, 0.0619]]).T

# Generar vector y, suponiendo la clases correcta es perro
# Para hacerlo vector columna se aplica la transpuesta usando .T
y = np.array([[0, 1, 0]]).T

# Matriz de pesos sinapticos, con valores aleatorios
W = np.random.randn (3, 4) * 0.01
# Comentar la línea anterior y descomentar las
# líneas siguientes en caso de querer replicar el
# mismo ejemplo que en el video
# W = np.array([[ 0.0032, -0.0266, 0.0006, -0.0097), 
#               [ 0.0036, -0.0101, -0.0081, 0.0084],
#               [-0.0006, 0.0050, -0.0065, 0.0265]])

# Vector de biases
b = np.zeros((3,1))

#Calcular scores
z =W@x + b

#Calcular prediccón
y_hat = Softmax(z)

# Calcular pérdida para el ejemplo, seleccionar valor donde y = 1 
loss_i= np.log(y_hat [np.where(y==1)])[0]

# Calcular pérdida para el ejemplo, seleccionar valor donde y = 1 loss_i = -np.log(y_hat [np. where (y==1)])[0]
# Obtener Gradientes dz, dw, y db
dz = y_hat - y
dW = dz@x. T
db = dz
# Actualizar parámetros 
learning_rate=1e-2
W = W - learning_rate*dW 
b = b - learning_rate*db
```
