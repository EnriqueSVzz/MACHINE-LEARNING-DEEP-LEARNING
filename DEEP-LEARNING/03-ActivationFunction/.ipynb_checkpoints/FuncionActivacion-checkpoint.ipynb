# Funciones de Activación

## Introducción a la Función de Activación

* Una `función de activación` es una función matemática que se aplica a la salida de una neurona:
$$
\begin{aligned}\\
z &= W \cdot x + b \\
a &= f(z)
\end{aligned}
$$
* Su trabajo es decidir `qué tanta información pasa a la siguiente capa`.
* ***¿Por qué son tan importantes?*** Sin funciones de activación:
  - Una `red neuronal` sería solo combinaciones lineales.
  - Aunque tengas muchas capas, `todo colapsaría a una sola ecuación lineal`.
  - `No podría aprender relaciones complejas`.

* Las funciones de activación introducen no linealidad, que es lo que permite:
  - Clasificar imágenes.
  - Reconocer voz.
  - Detectar fraude.
  - Aprender patrones complejos.

* ***Principales funciones de activación***:
$$
f(x)=
\begin{cases}
0, & x < 0 \\
1, & x \ge 0
\end{cases}
\quad \text{(Escalón)}
$$

$$
f(x)=x
\quad \text{(Lineal)}
$$

$$
f(x)=\frac{1}{1+e^{-x}}
\quad \text{(Sigmoide)}
$$

$$
f(x)=\tanh(x)=\frac{e^x-e^{-x}}{e^x+e^{-x}}
\quad \text{(Tangente hiperbólica)}
$$

$$
f(x)=\max(0,x)
\quad \text{(ReLU)}
$$

$$
f(x)=
\begin{cases}
x, & x \ge 0 \\
\alpha x, & x < 0
\end{cases}
\quad \text{(Leaky ReLU)}
$$

$$
f(x)=
\begin{cases}
x, & x \ge 0 \\
\alpha(e^x-1), & x < 0
\end{cases}
\quad \text{(ELU)}
$$

$$
f(x_i)=\frac{e^{x_i}}{\sum_j e^{x_j}}
\quad \text{(Softmax)}
$$

* `¿En qué paso entra la función de activación?`.
  - ***La función de activación siempre se aplica en el forward propagation, justo después de la combinación lineal***.

## Ejemplo Multicapa

* La principal razón de usar las funciones de activiación como y hemos visto, es que la salida se nuestra neurona será un valor lineal, mismo que si este lo mandamos a la segunda neurona no habrá gran impacto.
* A continuación deremos el ejemplo de lo que pasaría si no usamos una `función de Activación`.
* Empecemos defiendo nuestros valores en el primer `foward propagation`.

$$
\begin{aligned}
Z^{(1)} &= W^{(1)} \cdot X + b^{(1)} \\
Z^{(1)}_{(3,1)} &= W^{(1)}_{(3,4)} \cdot X_{(4,1)} + b^{(1)}_{(3,1)}
\end{aligned}
\quad \text{(Ec 0. Primera Red Neuronal)}
$$

* Ahora, según esto, nuestra segunda neurona recibiría los siguientes datos.
* Aquí nuestra `W2` ya es de ***3x3***, ya que la cantidad de neuronas son 3 y de entras también son 3.
* Nuestro `baias` siguen siendo ***3x1***.
$$
\begin{aligned}
Z^{(2)} &= W^{(2)} \cdot Z^{(1)} + b^{(2)} \\
Z^{(2)} &= W^{(2)}_{(3,3)} \cdot [W^{(1)}_{(3,4)} \cdot X_{(4,1)} + b^{(1)}_{(3,1)}] + b^{(2)}_{(3,1)}
\end{aligned}
\quad \text{(Ec 1. Segunda Red Neuronal)}
$$

* Ahora desarrollemos nuestra ecuación y veamos que pasa.
$$
\begin{aligned}
Z^{(2)} &= W^{(2)}_{(3,3)} \, W^{(1)}_{(3,4)} \, X_{(4,1)} + W^{(2)}_{(3,3)} \, b^{(1)}_{(3,1)} + b^{(2)}_{(3,1)} \\
Z^{(2)} &= W^{(3)}_{(3,4)} \, X_{(4,1)} +  (b_{(3,1)} + b^{(2)}_{(3,1)}) \\
\end{aligned}
\quad \text{(Ec 2. Obtención de los terceros valores)}
$$

* Podemos ver como nos da la misma formula, es decir `No cambios`.
