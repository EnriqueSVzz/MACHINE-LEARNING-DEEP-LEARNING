# Datos Categoricos Ordinales

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import MinMaxScaler
```

* Un conjunto de datos categoricos ordinales son un grupo de información los cuales guardan una relación entre sí.
* Existen 2 tipos de datos categóricos:
  * ***Nominales***: Grupos sin orden.
  * ***Ordinales***: Grupos con orden.

* Un ejemplo de grupo nominales, sería tener un conjunto de colores, ¿cómo se ordenaria estos datos?, realmente pareciera que no hay un indicador tan claro. No podemos establecer una orden.
* Para los ordinales, podemos ver el nivel de inglés de una persona, podemos clasificarlo como bajo, medio o intermedio.

### Datos Categóricos Ordinales

Aquí tenemos algunas clasificaciones de orden.
* Calificación Crediticia: [Mala, Regular, Buena Excelente]
* Encuestas de satisfacción: [ Muy Insatisfecho, Insastifecho, Nuetro,Satisfecho]
* Percepción de dolor: [Sin dolor, Dolor moderado, Severo]
* Podríamos establecer que son equidistante, es decir asignarles un peso para distancias.

```python
# Definimos nuestros clasificaciones
categorias_servicio = ["Muy insatisfecho", "Insatisfecho", "Neutral",
                       "Satisfecho", "Muy Satisfecho"]
categorias_calidad = ["Mala", "Buena", "Muy Buena", "Excelente"]

# Respuestas de algunos clientes
encuesta = {
    "servicio": ["Muy insatisfecho", "Insatisfecho", "Neutral",
                 "Satisfecho", "Muy Satisfecho", "Muy insatisfecho"],
    "alimentos": ["Mala", "Buena", "Muy Buena",
                  "Excelente", "Mala", "Buena"]
}

# 0: Cliente esperorádico 1: cliente frecuente
tipo_cliente = [0,0,1,1,0,1] # de los 6 clientes que opinarion
```

### Codificador Ordinal

* En este caso le pondrá una relación 1 a 1 a cada dato.
* Muy Insastifecho tendrá un cero ya que es el primer valor y mala por la primera también será igual.
* Está poniendo que son equidistante.
* En algunos casos esto no podría ser del todo correcto

```python
datos = pd.DataFrame(encuesta)

# Codificador ordinal con categorías definidas
codificador = OrdinalEncoder(categories=[categorias_servicio, categorias_calidad])

# Aplicamos codificación
datos = pd.DataFrame(
    codificador.fit_transform(datos),
    columns=["servicio", "alimentos"]
)

# Mostrar resultados
print(datos)
print(codificador.categories_)
```

### Codificador OneHotEncoder

* Este codificador los transforma en columnas, para que no sean equidistantes, al contrario veamos frecuencias. Es decir, para un registro podrá tener encendido y apagado.
* Cuando se ejecuta el `codificador.fit_transform(datos_one).toarray()`, las primeras 5 columnas son todos los tipos de "categorias_servicio" y los otros 4 son de categorias_calidad".
* Lo malo de esta forma de codificar, es que genera 1 columna por cada dato, por lo que en grandes conjuntos de datos, sería muy ineficiente.

```python
datos_one = pd.DataFrame(encuesta)
codificador = OneHotEncoder()
codificador.fit_transform(datos_one)

# Imprimimos los datos
#print(codificador.fit_transform(datos_one).toarray())
#print(codificador.categories_)
#print(np.concatenate(codificador.categories_))

datos_one = pd.DataFrame(
    codificador.fit_transform(datos_one).toarray(),
    columns=np.concatenate(codificador.categories_)
)
datos_one
```

### Comparación de las Codificaciones

* Clasificador: Regresión Logística
* Si hay diferencias a la hora de escojer un codificador.
* La probabilidad mayor, dictamina si es esporadico o frecuente.
* El MinMaxScaler de scikit-learn sirve para escalar (normalizar) tus datos numéricos a un rango específico, normalmente entre 0 y 1.
* Se utiliza en Machine Learning y análisis de datos para:
  * Evitar que una característica con valores muy grandes (como "ingresos") domine a otra con valores pequeños (como "edad").
  * Acelerar el entrenamiento de modelos.
  * Mejorar la precisión de algoritmos sensibles a la escala (como KNN, K-Means, Redes Neuronales, SVM).
  * Esto transforma todos los valores al rango [0, 1].
  * `MinMaxScaler(feature_range=(10, 20))`, podemos también cambiar el rango.

```python
print("\n*** Datos Escalados de la Codificación Ordinal")
escalador = MinMaxScaler()
print(datos)
datos = escalador.fit_transform(datos)
print("\n*** Datos Escalamiento de datos")
print(datos)

print("\n*** Clasificación con Datos codificados con OrdinalEncoder")
modelo = LogisticRegression().fit(datos,tipo_cliente)
print("Predicciones:", modelo.predict(datos))
print("Clases correctas:", tipo_cliente)
print("Calculo de Probabilidad\nSea Esporadico / Es un cliente frecuente\n", modelo.predict_proba(datos))

print("\n*** Clasificación con Datos codificados con OneHotEncoder")
modelo = LogisticRegression().fit(datos_one,tipo_cliente)
print("Predicciones:", modelo.predict(datos_one))
print("Clases correctas:", tipo_cliente)
print("Calculo de Probabilidad\nSea Esporadico / Es un cliente frecuente\n", modelo.predict_proba(datos_one))
```

### Ejemplo en Python: contraste con y sin MinMaxScaler

* x: El valor original que quieres escalar.
* x(min): El valor mínimo de la columna o serie.
* x(max): El valor máximo de la columna o serie.
* x(escalado): El nuevo valor, ahora normalizado entre 0 y 1.
$$
x_{escalado} = \frac{x - x_{\text{min}}}{x_{\text{max}} - x_{\text{min}}}
$$

```python

# Datos originales
df = pd.DataFrame({
    "salario": [1000, 2000, 3000, 4000, 5000],
    "edad": [22, 25, 30, 35, 40]
})

print("🔹 Datos originales:\n", df)

# Aplicamos MinMaxScaler
escalador = MinMaxScaler()
escalados = pd.DataFrame(escalador.fit_transform(df), columns=df.columns)

print("\n🔸 Datos escalados con MinMaxScaler:\n", escalados)
```
