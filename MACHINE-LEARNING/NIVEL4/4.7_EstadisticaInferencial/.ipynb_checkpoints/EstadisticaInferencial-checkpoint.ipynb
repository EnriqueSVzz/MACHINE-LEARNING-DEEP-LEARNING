### Estadística Inferencial

* La estadística se puede separar en 2 ramas:
  * Descriptiva: nos permite describir un conjunto de datos atraves de diferentes medidas, como la media, la moda, mediana, desviación estándar.
  * Inferencial: sacar conclusiones, pero usando un pequeño conjunto de datos.

* Población: todo nuestro conjunto de datos que podamos medir.
* Técnica de muestreo: forma de extraer aleatoriamente a un conjunto de datos para una muestra. Es muy importante que sea aleatorio, para evitar sesgos. Es como si solo escogieramos de un país para hablar de todo el mundo.
* Media muestral: promedio de la muestra
* Desviación muestral.
* varianza muestral.
* Con esta muestra podemos ver o predecir para la población.
* Las pruebas de hipotesis. La altura de europeos es mayor a la de los asiaticos.
