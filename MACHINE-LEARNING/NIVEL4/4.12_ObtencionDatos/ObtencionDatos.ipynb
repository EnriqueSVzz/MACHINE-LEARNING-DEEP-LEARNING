### Cargar Datos de Juguete

```python
# Cargamos el primer conjunto de llaves
diabetes = datasets.load_diabetes(as_frame=True)
# Imprimimos
for llave in diabetes.keys():
    print(diabetes[llave], "\n"*5)
```

```python
digitos = datasets.load_digits()
print(digitos.keys())
print(digitos["target"][0])
plt.matshow(digitos["images"][0])
print(digitos["target"][5])
plt.matshow(digitos["images"][5])
plt.show()
```
