# Medidas de Dispersión

* Rango
* Varianza
* Desviación Estándar
* Rango Intercuartílico

### Noción de Dispersión

Muestra A = [86,89,87,88,86,88,89,87,87]
* Pesos similares
* Dispersión baja, todos estan muy parecidos.

Muestro B = [170,5,90,25,35,55,15,40,80]
* Pesos variados
* Mucha dispersión

### Rango,  Varianza

* ***Rango***: de la muestra B, Máximo(170) - Minimo(5) = 165
* ***Varianza Poblacional***: primero calculamos el promedio,a cada punto le restamos el promedio y el resultado lo elevamos al cuadrado, ya que solo queremos ver que tanto varian del promedio y dividimos sobre el total de productos.
* ***Varianza Muestral***: es lo mismo se suman las diferencia al cuadrado y se divide entre N-1, esto ya que hay que darnos un margen de error.
* ***Desvición Estándar Poblacional***: es sacar la raiz cuadrada a la Varianza Poblacional, ya que las unidades de la varianza estarían también al cuadrado como m2.
* ***Desviación Estándar Muestral***: es sacar la raíz cuadrada de la Varianza Muestral
* ***Rango Intercuartílico***: los datos tienen que estar ordenados, dividirlo en 4 partes, y sería la resta del Cuartil 3 - Cuartil 1, se usa para calcular los rangos atípicos.

### Medidas de Dispersión con Numpy

* `ddof=1` grados de libertad.

```python
import numpy as np

# Muestra B ordenada
pesos = [5,15,25,35,40,55,80,90,170]

# Imprimos los rangos
print("Rango-> MAX[", np.max(pesos), "], MIN[", np.min(pesos), "]")

# Varianzas
print("Varianza poblacional:", np.var(pesos))
print("Varianza muestral:", np.var(pesos, ddof=1))

# Desviacion
print("Desviacion Est poblacional:", np.std(pesos))
print("Desviacion Est muestral:", np.std(pesos, ddof=1))

# Calculo del rango intercuartílico
Q1 = np.percentile(pesos, 25)
Q3 = np.percentile(pesos,75)
print("Rango Intercuartilico:", Q3-Q1)
```
