# Datos Categoricos

* Tenemos 2 tipos de datos: númericos y categoricos
* Dato categoricos: país en el que nacemos, tipo de sangre etc.
* Para el aprendizaje de máquinas estos datos pueden ser un reto.
* Tenemos que hacer una transformación para cambiar estos datos a númericos, ya que por ejemplos las redes neuronales utilizan estos datos.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
```

### Contexto y Datos Categóricos

```python
datos = {
    "nombre": ["Mariana", "Ana", "Elsa", "Gustavo", "Pedro", "Raul", "Carlos", "José", "Luis"],
    "saldo": [10000.00, 8000.00, 9000.00, 2000.00, 2100.00, 12000.00, 5000.00, 10000.00, 200.00],
    "pais": ["Argentina", "Bolivia", "Chile", "Colombia", "Costa Rica", "Ecuador", "Mexico", "Peru", "Peru"]
}

datos = pd.DataFrame(datos)

# Cambiamos el tipo de datos de object a Category
datos["pais"] = datos["pais"].astype("category")
datos.info()
```

### Inapropiada Codificación con Reemplazo

* Con estos valores, la máquina podría tomarlos como proximidad entre paises o referencia a su tamaño, población etc.
* Esto no se debe hacer y podríamos meter un sesgo a nuestra información.

```python
datos_sesgados = datos.copy()
reemplazos = {
    "Argentina":1, 
    "Bolivia":2,
    "Chile":3, 
    "Colombia":4, 
    "Costa Rica":5, 
    "Ecuador":6, 
    "Mexico":7, 
    "Peru":8
}
# Esto para los datos que no son category
#datos_sesgados["pais"].replace(reemplazos)
datos_sesgados["pais"] = datos_sesgados["pais"].cat.rename_categories(reemplazos)
datos_sesgados
```

### Codificación de Categorías One-Hot

* El primer 1 nos esta indicando que el primer objeto pertenece a Argentina
* El segundo 1, también hace referencia a otros país.
* Al final vemos que perú venía 2 veces, por lo que podemos ver doble 1 en la columna para Peru.
* Filas = clientes, Columna=Pais
* El 1 hace referencia a que esta encendido.

```python
codificador = OneHotEncoder()
codificacion = codificador.fit_transform(datos[["pais"]])
print(type(codificacion))
print(codificacion)
print(codificacion.toarray())

nuevas_cols = pd.DataFrame(codificacion.toarray(),
                          columns=codificador.categories_)
print(nuevas_cols)
datos = pd.concat(
    [datos,nuevas_cols],
    axis="columns"
)
datos
```

```python
datos.drop("pais", axis="columns", inplace=True)
datos
```
