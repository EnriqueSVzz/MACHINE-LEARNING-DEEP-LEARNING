# Correlación Spearman

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats
```

* La correlación de Spearman, también conocida como rho de Spearman (ρ), es una medida no paramétrica que evalúa la fuerza y dirección de la relación entre dos variables usando rangos en lugar de los valores originales.
* ¿Cuándo se usa Spearman en vez de Pearson?
  * Pearson: Relación lineal, Valores reales, Requiere distribución normal (idealmente).
  * Spearman: Relación monótona (creciente o decreciente), Rangos (posiciones ordenadas), No requiere normalidad.

$$
\rho = 1 - \frac{6 \sum d_i^2}{n(n^2 - 1)} \quad \text{donde:} \quad
\begin{cases}
d_i = \text{diferencia entre los rangos de } x_i \text{ e } y_i \\
n = \text{número total de pares}
\end{cases}
$$

* Mide la fuerza y dirección de asociación entre 2 variables ordenadas.
* Mide la correlación entre rangos (o rankings) de 2 variables.
* Mide la monotonicidad de la relación entre 2 variables.
* Posibles valores de -1 a 1.
* Tienen que estar ordenados para su comparación.

### Reglas Monotónicas

* Positivas: si una variable crece, la otra también.
* Negativa: si una descrece la otra también.
* No Monotónicas: una rho cercana a o cero, en la gráfica se ve dispersa o curvas.
* Monotónico vs Lineal: 

### Monotónico vs Lineal

* Cuando es una linea, ambas correlaciones puede dar 1, en este caso no hay diferencias.
* En caso que la gráfica este un poco curveada, Pearson podría dar un valor cercano a 1, mientras que Spearman seguiría dando 1.
* La correlación Spearman es más flexible, ya que va por rangos.
* En Spearman, mientras las 2 variables aumenten, seguirá marcando 1.
* Caso contrario seguirá marcando -1.

```python
goles = [2,15,3,8,0,11]
asistencias = [4,7,8,6,1,13]
print(stats.spearmanr(goles,asistencias))
print(stats.spearmanr(goles,asistencias)[0])
```
