# Discretizar Datos Númericos

```python
import numpy as np
import pandas as pd
```

### ¿Qué son?

* Son el proceso contrario a la clasificación, es decir tomar una variable categorica y convertilo en en una variable númerica continua.
* Esto sirve para que en lugar de tener un modelo de regresión lineal, tengamos un modelo de clasificación.
* Imagina que tenemos un dataset de temperaturas, con este proceso quiza podríamos tener: frio, tibio y caliente.
* Esto también podría eliminar los outliers, ya que ahora quedarían en un bloque generico, ya sea "temperatura muy alta" o "baja".
* Al proceso de Discretizar también se le conoce como Compartimentar y Categorizar.
* Dentro de los datos númericos hay 2 tipos de datos Continuos y Discretos.
* Datos Númericos Continuos:
  - Pueden tomar cualquier valor dentro de un rango determinado, incluyendo decimales.
  - Se obtienen normalmente mediante mediciones.
  - Entre dos valores siempre puedes encontrar infinitos valores intermedios.
  - La altura de una persona (1.75 m, 1.751 m, etc.)
  - El peso de un objeto (70 kg, 70.5 kg)
  - La temperatura (22.3 °C, 22.31 °C)
* Datos numéricos Discretos:
  - Solo pueden tomar ciertos valores específicos, generalmente enteros.
  - Se obtienen mediante conteo.
  - No hay valores intermedios entre dos números adyacentes.
  - Número de hijos en una familia (0, 1, 2, 3…)
  - Cantidad de autos en un estacionamiento (5, 6, 7…)
  - Veces que una persona ha ido al cine este mes

### Categorizar

* Imagina que tenemos las siguientes edades
* 1, 7, 8, 15, 16, 28, 35, 50, 55, 70, 75, 100
* Podríamos clasificarlos en 3 categorías: Baja, Media, Alta
* Para asignar categorías podríamos hacer lo siguiente:
* [0.901, 34], (34,67], (67, 100] // Separado en 3, rangos de 34 en 34
* (1, 7, 8, 15, 16, 28) ; (35, 50, 55) ; (70, 75, 100)
* No es en base a la cantidad de elementos.
* Es en base a los rangos de los datos.

### Discretización

* Sin el `include_lowest` no incluye el valor más bajo.

```python
edades = np.array([1,7,8,15,16,28,35,50,55,70,75,100])
resultado = pd.cut(
    edades,
    bins=3,
    labels=["baja", "media", "alta"],
    include_lowest=True,
    retbins=True
)
#resultado
print(resultado, "\n")

# Esto calculo los rangos
print(resultado[1], "\n")

# Clasificaciones aplicadas
print(resultado[0].codes, "\n")

# Tipos de Categorias
print(resultado[0].categories, "\n")

# Valores convertidos.
print(np.array(resultado[0]), "\n")
```

### Contenedores de Diferentes Tamaños

* Imagina que tenemos las siguientes edades
* 1, 7, 8, 15, 16, 28, 35, 50, 55, 70, 75, 100
* Podríamos clasificarlos en 4 categorías: Infante, Joven, Adulto y Mayor.
* [0, 11], (11,17], (17,59], (59, Infinito).
* La diferencia del anterior, es que ahora nosotros establecemos los rangos.
* Para especificar el infinito, usamos el `np.inf`, ya que es el número más grande que numpy.

```python
edades = np.array([1,7,8,15,16,28,35,50,55,70,75,100])
resultado = pd.cut(
    edades,
    bins=[0, 11, 17, 59, np.inf],
    labels=["Infante", "Joven", "Adulto", "Mayor"],
    include_lowest=True,
    retbins=True
)
#resultado
print(resultado, "\n")

# Esto calculo los rangos
print(resultado[1], "\n")

# Clasificaciones aplicadas
print(resultado[0].codes, "\n")

# Tipos de Categorias
print(resultado[0].categories, "\n")

# Valores convertidos.
print(np.array(resultado[0]), "\n")
```

### Ejemplo

* Vamos a Discretizar el genero en "Masculino y Femenino". Aquí no es necesario establecer rangos.
* Discretizar la variable de salida (target), que representa la progresión de la enfermedad un año después, y convertirla en un problema de clasificación.

```python
from sklearn.datasets import load_diabetes
import pandas as pd

# Cargar datos
diabetes = load_diabetes()
X = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)

# Discretizar 'sex' en 2 categorías
resultado = pd.cut(
    X["sex"],
    bins=2,
    labels=["Masculino", "Femenino"],
    include_lowest=True,
    retbins=True
)

# Mostrar los rangos de corte
print("Rangos de corte:", resultado[1], "\n")

# Mostrar las etiquetas como texto ("Masculino", "Femenino")
#print("Valores categóricos:", np.array(resultado[0]), "\n")

# Si quieres agregarlos al DataFrame:
X["sexo_cat"] = resultado[0]
print(X[["sex", "sexo_cat"]].head())
```

```python
y = diabetes.target  # Esta es la variable de regresión
# Discretizar y en 3 clases: Bajo, Medio, Alto
y_cat = pd.qcut(y, q=3, labels=["Bajo", "Medio", "Alto"])
print(y_cat.value_counts())
X["progresion_cat"] = y_cat
print(X[["progresion_cat"]].head())
```
