# Correlación Pearson y Matriz de Correlaciones

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

* La correlación de Pearson (también llamada coeficiente de correlación de Pearson o simplemente r) es una medida estadística que indica el grado de relación lineal entre dos variables cuantitativas.
* ¿Qué mide?
  * Qué tan fuerte y en qué dirección se relacionan dos variables numéricas.
  * Solo mide relaciones lineales (no curvas o no lineales).

$$
r = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}
{\sqrt{\sum_{i=1}^{n} (x_i - \bar{x})^2} \cdot 
\sqrt{\sum_{i=1}^{n} (y_i - \bar{y})^2}} \quad
\text{donde:} \quad
\begin{cases}
x_i, y_i \text{ son los valores individuales} \\
\bar{x}, \bar{y} \text{ son las medias de } x \text{ e } y
\end{cases}
$$


r	Interpretación
* +1	Correlación lineal perfecta positiva
* 0	Sin correlación lineal
* –1	Correlación lineal perfecta negativa
* Entre –1 y 1	Correlación parcial (más cerca a ±1, más fuerte)

### Magnitud y Dirección de una relación lineal

```python
personas = pd.read_csv("personas.csv")
personas
```

### Relación Altura y Peso

* Podemos ver que hay una especie de relación lineal, a mayor altura mayor peso en nuestros datos.
* Es una tendencia positiva.

```python
plt.scatter(x=personas["altura"], y=personas["peso"], color="green")
plt.ylabel("Peso (kgs)")
plt.xlabel("Altura(cms)")
plt.show()
```

### Relación Ingreso y Horas Trabajadas

* No se ve una relación tan lineal, solo se ve una mancha negra.
* Mientras más hiras trabajas, no siempre se gana más ingreso.

```python
plt.scatter(x=personas["ingreso"], y=personas["horas_trabajadas"], color="pink")
plt.xlabel("Ingreso ($)")
plt.ylabel("Horas trabajadas")
plt.show()
```

### Relación Ingreso y Ausencias

* Se ve que hay una especie de relación, a menor ausencias más ingreso.
* Relación Negativa

```python
plt.scatter(x=personas["ingreso"], y=personas["ausencias"], color="red")
plt.xlabel("Ingreso ($)")
plt.ylabel("Ausencias")
plt.show()
```

### Relación Ingreso y Peso de una persona

* No se ve que haya relación, solo una mancha.
* El peso no influye en la cantidad de tus ingresos.

```python
plt.scatter(x=personas["ingreso"], y=personas["peso"], color="purple")
plt.xlabel("Ingreso ($)")
plt.ylabel("Peso")
plt.show()
```

### Matriz de Correlación

* `corr()`: nos trae todos los coeficientes de correlación, entre las combinaciones de sus columnas.
* -1: existe una relación fuerte no negativa.
* 1: existe una relación fuerte positiva.
* 0: no hay relación.
* Valores Intermedio (-1,1): que tan fuerte o débil son.
* La diagonal siempre será un 1, ya que es misma fila-columna.
* `matshow()`: podemos establecer los colores que vaya de un blue, pasa por white y termina en red.

```python
matriz = personas.corr()
plt.matshow(matriz, cmap="bwr", vmin=-1, vmax=1)

# ponemos las columnas
plt.xticks(range(5), personas.columns, rotation=90)
plt.yticks(range(5), personas.columns)

# Imprimos el contenido de cada casilla
for i in range(len(matriz.columns)):
    for j in range(len(matriz.columns)):
        plt.text(i,j, round(matriz.iloc[i,j],2), ha="center", va="center")

plt.colorbar()
plt.show()
```

### Calculo del coeficiente Manual

```python
# Correlación de Pearson
correlacion = personas['peso'].corr(personas['altura'], method='pearson')
print(f"Coeficiente de correlación de Pearson: {correlacion:.2f}")
```
