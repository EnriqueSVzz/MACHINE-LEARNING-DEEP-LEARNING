# Imputación (Manejo de Datos Faltantes)

```python
import pandas as pd
```

### Manejo de Datos Faltantes

* Los datos faltanes serán marcados como NaN
* NaN significa ***Not a Number***

```python
datos = pd.read_csv("usuarios.csv")

# Imprimimos todos los datos
print(datos["nombre"])

# Este nos marcará todos los NaN
print("\n", datos["nombre"].isnull())

# Podemos hacer una busqueda de los valores que cumplen con el nulo en el nombre
print("\n", datos[ datos["nombre"].isnull()] )

# Dropear todos los valores nulos
print("\n", datos.dropna())

# Igualmente podemos borrar por columnas deseada
print("\n", datos.dropna(subset=["nombre","ingreso"]))
```

### Modifición de Datos

```python
# Esto ya no va a regresar un dataframe
datos.dropna(subset=["nombre","ingreso"], inplace=True)
print("\n", datos)
```

* También aquí podríamos usar el inplace

```python
# Por su lado aquí podemos llenar NaN con ceros.
print("\n", datos.fillna(0))

# Igualmente podríamos poner otro valor por columna
print("\n", datos["nombre"].fillna("Anonimous"))

# Establecer los valores default en NaN
valores_por_defecto = {
    "nombre": "DESCONOCIDO",
    "edad": 0,
    "ingreso": 0
}
print("\n", datos.fillna(value=valores_por_defecto))
```

### Calculos con NaN

```python
promedio = datos["ingreso"].mean()
mediana = datos["ingreso"].median()
moda = datos["ingreso"].mode()
print(promedio,mediana,moda)

datos = datos["ingreso"].fillna(promedio)
print("\n", datos)
```
