# Tuplas

## Declaración

```python
### No podemos modificar las tuplas, pero es mucho más rápida y protege los datos
### En si es lo mismo que una lista, pero estatica
tupla = ("Juan", 22)
print("\n Tupla[0]: ", tupla[0])

### Esto no se puede hacer, a menos que el objeto adentro sea modificable.
# tupla[0] = "Juana"
```

## Asignación

```python
### Podemos transformar una tupla a lista.
lista = list(tupla)
print("\n lista", lista)

### E igualmente regresar a una tupla 
lista.append("México")
tupla1 = tuple(lista)
print("\n Tupla1: ", tupla1)

### Igualmente podemos convertirla en conjunto.
### Recordemos que en el conjunto no hay garantía de orden
conjunto = set(tupla1)
print("\n Conjunto: ", conjunto)
```
