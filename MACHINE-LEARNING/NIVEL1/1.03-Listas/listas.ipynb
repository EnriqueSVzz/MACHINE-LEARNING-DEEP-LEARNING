# Listas

## Declaración de una lista

```python
edades = [13, 14,34,45,65]
print("Edad[2]: ", edades[2])
```

## Asignación a un elemento

```python
edades[2] = 23
print("Nuevo Edad[2]", edades[2])
```

## Sliding

```python
#Sublista (el ultimo no esta considerado)
print("Edad {0,2}", edades[0:2])
print("Edad {:4}", edades[:4])
```

## Funciones sobre listas

```python
print("Len de Edad: ", len(edades))
print("Edad Maxima ", max(edades))
print("Edad Mimina ", min(edades))
print("Suma de todas las edades ", sum(edades))
print("Promedio las edades ", sum(edades)/len(edades))
```

## Operaciones sobre listas

```python
### Insertando un valor en la posición [2]
print("Vieja Lista: ", edades)
edades.insert(2, 50)
print("Nueva Lista: ", edades)
```

```python
### Agregar un valor al final de la lista
print("Lista: ", edades)
edades.append(45)
print("Agregando con Append al Final: ", edades)
```

```python
print("Lista: ", edades)
edades.extend([12,8,24,50,66])
print("Concatenando Listas:", edades)
```

```python
### Borrando un dato por index
print("Lista: ", edades)
del edades[1]
print("Lista despues del borrado:", edades)
```

* `remove(valor)`: elimina el primer elemento cuyo valor coincida con el que indiques, no el elemento en esa posición.
* Si hubiera más, elimina al primero que se encuentre recorriendo.

```python
### Borrando un dato por valor
print("Lista: ", edades)
edades.remove(45)
print("Lista despues del borrado:", edades)
```

```python
### Pop
print("Pop:", edades.pop())
```

```python
### Frecuencias
print("Frencuencia de 50:", edades.count(50))
```

```python
### Buscar un index
print("Index de 50:", edades.index(50))
```

```python
### Lista Ordenada (Mayor a menor)
### Default es de menor a mayor
### Estas funciones no guardan nada, hay que accionarlas.
edades.sort(reverse=True)
print("Lista Ordenada:", edades)
edades.reverse()
print("Lista Ordenada:", edades)
```
