# Funciones

## Definiciones básicas

```python
def suma(a, b):
    suma = a+b
    return suma

def addByList(lista):
    total = 0
    for i in lista:
        total +=i
    
    return total
```

## Funciones más elaboradas

* El `*` hace que todos los argumentos que pases se agrupen en una tupla llamada `numeros`.

```python
### Sumar por 'N' cantidad de argumentos
def addByArgs(*numeros): 
    total = 0
    for i in numeros:
        total +=i
    return total
```

* `.capitalize()`: primera letra mayuscula y las demás minusculas.

```python
### Función que da formato a los string
def getFormattedString(nombre, apePat, apeMat):
    formato = nombre.capitalize() + " "
    formato += apePat.capitalize() + " "
    formato += apeMat.capitalize()
    return formato
```

* `strip()`: Elimina los espacios en blanco al inicio y al final del texto.
* No elimina los espacios del medio, solo los de los extremos.
* `split()`: Divide el texto en una lista usando un separador.
* Si no se pasa nada, separa por espacios.

```python
# Función que Calcula el RFC
def getRFC(nombre, fechaNac, codigo):
    nombre = nombre.upper()
    partes = nombre.strip().split()  # Divide en ['JUAN', 'PEREZ', 'LOPEZ']
    
    a_paterno = partes[1]
    a_materno = partes[2]
    nombre_persona = partes[0]

    # Buscar primera vocal interna del apellido paterno
    vocal = ''
    for letra in a_paterno[1:]:
        if letra in 'AEIOU':
            vocal = letra
            break

    RFC = a_paterno[0] + vocal
    RFC += a_materno[0]
    RFC += nombre_persona[0]

    # Agregar fecha en formato: AAMMDD
    RFC += fechaNac[8:]    # Año: "1988"
    RFC += fechaNac[3:5]   # Mes: "05"
    RFC += fechaNac[:2]    # Día: "14"

    RFC += codigo

    return RFC
```
