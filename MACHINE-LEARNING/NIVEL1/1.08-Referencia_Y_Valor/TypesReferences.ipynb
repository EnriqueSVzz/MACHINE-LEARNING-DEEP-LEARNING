# Tipos de Referencias

* Los Tipos de Datos primitivos se asumen que serán pequeños.
* Por consiguiente, se pasarán por valor.
* Las estructuras se espera que tengan un tamaño mucho más grande, por lo que serán enviado por referencia.

```python
def doblar(referencia, valor):
    referencia *=2
    valor *=2
    print("\n Dentro de la Función: ", referencia, valor)
    
    
struct = ["a", "b", "c"]
primi = "abc"

print("\n Antes: ", struct, primi)
doblar(struct, primi)
print("\n Despues: ", struct, primi)
```
