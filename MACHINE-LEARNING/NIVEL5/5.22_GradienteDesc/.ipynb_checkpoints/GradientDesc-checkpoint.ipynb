# Gradiente Descendiente

### ¿Qué es?

* El `Gradiente Descendiente (Gradient Descent)` es un `algoritmo de optimización` que se usa para `ajustar los parámetros (pesos y sesgos)` de una red neuronal con el fin de `minimizar una función de error o pérdida (loss function)`.
* Es un método para encontrar los valores de los parámetros que hacen que la red cometa el menor error posible.
