# Técnica de Validación Cruzada

### ¿Qué es?

* Una de los aspectos más importantes no es sobrestimar las capacidades de mi modelo.
* Está es una de las técnicas más robustas para evaluar modelos de aprendizaje tanto de Regresión como Clasificatorios.
* Entrenar con el 100% de los datos, no tiene chiste ya que el modelo ya conocería la respuesta, por esto mismo existe el set de prueba.
* Los datos de validación, estos nos permiten ajustarlos hiperparametros de un modelo. Por ejemplo en una red nueronal, sería la cantidad de neuronas y capas. Esto evita que el modelo se haga muy bueno para un tipo de caso.
* También puede pasar que dentro de los datos de Pruebas no haga mucha complejitud y sean predicciones más comunes.

### Evaluación Cruzada

* Aquí principalmente los datos son divididos o n pliegues.
* En este ejemplo son 5 pliegues, los primeros 4 son utilizados para entrenamiento y el ultimo para evaluar el modelo, pero esto solo para la primera iteración
* Para la siguiente iteración, se utilizará el 4 pliegue.
* Para el n-1 hasta que llega al primer conjunto.
* Finalmente de las K iteraciones, se cada un promedio.
* Entre 5 a 10 pliegues podría ser los más correcto.
* Si llegamos a tener un conjunto masivo de datos, podríamos tener un solo conjunto especializado.

### Evaluación de Modelos

* No importa el tipo de modelo, aplica para cualquier tipo de modelo.
* `[0.75 0.9  0.7  0.55 0.8 ]`: esto claramente representa que por datasets hubo unos más faciles de indentificar por esos.

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

personas = pd.read_csv("salario.csv")

### Practicamente aquí ya le pasamos todas las respuestas
bosque = RandomForestClassifier()
bosque.fit(
    personas[["edad", "estudio"]].values,
    personas["ingreso"].values
)

### Evaluación Ingenua
print(bosque.score(
     personas[["edad", "estudio"]].values,
    personas["ingreso"].values
))

### Esto regresa una arreglo de numpy con las evaluaciones
print(cross_val_score(
    bosque,
    personas[["edad", "estudio"]].values,
    personas["ingreso"].values,
    cv=5
))

### promedio
print(cross_val_score(
    bosque,
    personas[["edad", "estudio"]].values,
    personas["ingreso"].values,
    cv=5
).mean())
```
