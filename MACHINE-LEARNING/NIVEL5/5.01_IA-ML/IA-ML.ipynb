# Inteligencia Artificial y Aprendizaje de Máquina

### ¿Qué son?

* El Aprendizaje de Máquina es un subconjunto de la Inteligencia Artificial.

### Inteligencia Artificial.

* Es un campo amplio de la informática que busca crear sistemas capaces de realizar tareas que normalmente requieren inteligencia humana.
* Estas tareas pueden incluir: razonamiento, resolución de problemas, comprensión del lenguaje, visión, planificación, etc.

* ***La IA es el “todo”, el objetivo general***.

### Machine Learning

* Es una subdisciplina dentro de la IA.
* Se centra en crear algoritmos que permiten a las máquinas aprender de los datos y mejorar su rendimiento con el tiempo sin ser explícitamente programadas.
* En lugar de decirle a la máquina qué hacer paso a paso, se le dan datos y ella “aprende” patrones.
* ***El ML es una técnica para lograr IA***.
* `IA`: Quieres crear un coche autónomo que conduzca solo.
* `ML`: Para que el coche aprenda a reconocer peatones y semáforos, usas algoritmos de aprendizaje automático con miles de imágenes de tráfico.
* ***En Machine Learning, el tipo de problema no depende tanto del tipo de datos de entrada, sino del tipo de salida (target) que queremos predecir.***

### Técnica Supervisada

* Es cuando el modelo aprende a partir de datos etiquetados, es decir, cada entrada ***ya tiene su respuesta correcta***.
* Problemas comunes:
  - Clasificación: predecir categorías (ej. gato vs. perro).
  - Regresión: predecir valores numéricos (ej. precio de una casa).
* Ahora que dada una combinación de las variables, podemos clasificar en 2 tipos reacciones o emociones 'Positiva' y 'Negativa'.
* Podríamos tener una tabla con algunos emoticones para que el modelo pueda diferenciar.
* Entonces el objetivo del modelo es que sea capaz de identificar y clasficarlas.
* Entradas (features): variables de entrada (por ejemplo, edad, tamaño, color...).
* Etiquetas (targets): lo que queremos predecir (por ejemplo, precio, categoría, etc.).

### Técnica No Supervisada

* Problemas comunes:
  - Clustering (agrupamiento): encontrar grupos naturales (ej. segmentar clientes).
  - Reducción de dimensionalidad: simplificar los datos sin perder la información importante (ej. PCA).
* Aquí los datos no están etiquetados. El modelo intenta encontrar patrones ocultos o estructuras por sí solo.
* Estas técnicas se utilizan para entender relaciones entre las observaciones o variables.
* Las observaciones ***NO*** tienen una respuesta asociada que guíe el aprendizaje.
* Aprenden con cosas que no habían tenido antes.
* Imagina que nuestro modelo se le presenta un nuevo color que no tenía contemplado y es clasificado como "otro".
* PCA = Principal Component Analysis. Es una técnica de reducción de dimensionalidad usada en Machine Learning y análisis de datos.
  - Cuando tienes muchas variables o características (por ejemplo, 100 columnas en un dataset), puede volverse difícil de analizar o visualizar.
  - PCA toma esos datos complejos y los transforma en una versión más simple, con menos dimensiones, pero conservando la mayor parte de la información importante.

### Deep Learning (DL) — Aprendizaje Profundo

* Es una subárea del Aprendizaje Automático (ML) que usa redes neuronales artificiales profundas, inspiradas en cómo funciona el cerebro humano.

* Funciona especialmente bien con grandes cantidades de datos y tareas como reconocimiento de imágenes, voz y texto.
* Es como un cerebro artificial que aprende viendo ejemplos miles (o millones) de veces hasta que puede hacerlo por sí solo.

  - Reconocer rostros en fotos (como lo hace Facebook o Google Fotos).
  - Detectar objetos en un video.
  - Generar imágenes (como DALL·E).
  - Conducir coches autónomos (reconocer señales, peatones, etc.).

### NLP (Natural Language Processing) — Procesamiento del Lenguaje Natural.

* Es una rama de la IA (que también usa ML y DL) que permite a las máquinas entender, interpretar, generar y responder en lenguaje humano.
* Incluye tareas como entender frases, traducir idiomas, responder preguntas, etc.
* Ejemplos:
  - Chatbots.
  - Traducción automática (Google Translate).
  - Análisis de sentimientos (por ejemplo, saber si una reseña es positiva o negativa).
  - Corrección automática de texto (como Grammarly o el corrector de Word).

# Regresión vs Clasificación.

### 1. Regresión

* Regresión-Cuantitativo.
* Problemas con una respuesta cuantitativa, son problemas de regresión.
* Se dice que tenemos una Regresión, si queremos predicir una variable cuantitativa.
* Por ejemplo el tiempo el tiempo que tarda un corredor en correr 100 metros en base a las horas de entrenamiento.
* Principalmente en estos problemas se comienza por una visualización de los datos y tratar de encontrar un patrón.
* Probablemente en este caso mientras más horas hayas entranado menor será tu tiempo.
* Modelos comunes:
  - `Regresión lineal`.
  - `Regresión polinómica`.
  - `Árboles de decisión para regresión`.
  - `Random Forest Regressor`.
  - `Redes neuronales regresoras`.
* Métricas comunes:	`MAE`, `MSE`, `R²`.
* ***MAE (Mean Absolute Error)***: error promedio absoluto.
* ***MSE (Mean Squared Error)***: error promedio al cuadrado.
* ***R² (Coeficiente de determinación)***: qué tan bien explica el modelo la variabilidad.

### 1.1 La ecuación básica

*Para un solo feature (regresión lineal simple), la relación es:*
$$
y = \beta_0 + \beta_1 x + \varepsilon
$$
donde:
\[
\begin{aligned}
y & = \text{variable que queremos predecir (target)} \\
x & = \text{variable de entrada (feature)} \\
\beta_0 & = \text{intercepto (valor de } y \text{ cuando } x=0) \\
\beta_1 & = \text{pendiente (cambia } y \text{ por cada unidad que cambia } x) \\
\varepsilon & = \text{error (diferencia entre predicción y realidad)}
\end{aligned}
\]

* El algoritmo trata de encontrar los valores de 𝛽0 y 𝛽1 que minimizan el error total (usualmente el error cuadrático medio, MSE).
* Si tenemos puntos en un plano (𝑥,𝑦), la regresión lineal ajusta la mejor línea recta que pase lo más cerca posible de todos esos puntos.

*  Cálculo de la pendiente 𝛽1
$$
\beta_1 = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}{\sum_{i=1}^{n} (x_i - \bar{x})^2}
$$

* Cálculo del intercepto 𝛽0
$$
\beta_0 = \bar{y} - \beta_1 \bar{x}
$$


* Cálculo de la predicción $$\hat{y}_i$$
$$
\hat{y}_i = \beta_0 + \beta_1 x_i
$$

* Cálculo del error o residuo 𝜀𝑖
$$
\varepsilon_i = y_i - \hat{y}_i
$$


* Función de costo (MSE - Error Cuadrático Medio)
$$
J(\beta_0, \beta_1) = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

*  Coeficiente de determinación 𝑅2
$$
R^2 = 1 - \frac{\sum (y_i - \hat{y}_i)^2}{\sum (y_i - \bar{y})^2}
$$

### 1.2 Cálculo manual de 𝛽0 y 𝛽1 en Python

* `beta_1`: literal es un número que es la división de 2 sumatorias.
* `beta_0`: literal es solo es la multiplicación usando beta1.
* Las predecciones es beta1*x + beta0.
* Y el error es el valor real - la predicción.

```python
import numpy as np

# Datos
X = np.array([1, 2, 3, 4, 5])
y = np.array([3, 4, 2, 5, 6])

# 1. Calcular medias
x_mean = np.mean(X)
y_mean = np.mean(y)

# 2. Calcular beta_1 (pendiente)
numerador = np.sum((X - x_mean) * (y - y_mean))
denominador = np.sum((X - x_mean)**2)
beta_1 = numerador / denominador

# 3. Calcular beta_0 (intercepto)
beta_0 = y_mean - beta_1 * x_mean

print(f"beta_1 (pendiente): {beta_1}")
print(f"beta_0 (intercepto): {beta_0}")

# 4. Calcular las predicciones
y_pred = beta_0 + beta_1 * X

# 5. Calcular los residuos (errores)
residuos = y - y_pred

print(f"Predicciones: {y_pred}")
print(f"Residuos: {residuos}")
```

### 1.3 Evaluación del Modelo de Regresión

* Una vez que entrenamos el modelo y obtenemos predicciones 𝑦^, necesitamos medir qué tan lejos están del valor real 𝑦.

 1. ***Error absoluto medio (MAE):***
    * Promedio de la magnitud de los errores (sin importar si están por encima o por debajo):
      $$
      \text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|
      $$
    * Fácil de interpretar.
    * Se mide en las mismas unidades que 𝑦.
    * No penaliza mucho los errores grandes.

2. ***Error cuadrático medio (MSE):***
   $$
   \text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
   $$
   * Penaliza más los errores grandes.
   * Es sensible a outliers.

3. ***Raíz del error cuadrático medio (RMSE):***
   $$
   \text{RMSE} = \sqrt{\text{MSE}}
   $$
   * Tiene las mismas unidades que el target.
   * Más interpretable que el MSE.

4. ***Coeficiente de determinación 𝑅2***
   $$
   R^2 = 1 - \frac{\sum (y_i - \hat{y}_i)^2}{\sum (y_i - \bar{y})^2}
   $$
   * Mide qué proporción de la varianza de los datos es explicada por el modelo.
   * Valores:
     - `𝑅2=1`: el modelo predice perfectamente.
     - `𝑅2=0`: el modelo no explica nada.
     - `Puede ser negativo si el modelo es peor que usar la media`.

***Interpretación***

1. MAE: 0.64
   * El error promedio absoluto entre las predicciones y los valores reales es 0.64 unidades.
   * Ejemplo: si estuvieras prediciendo precios en miles de dólares, el modelo se equivoca en promedio unos 640 dólares.


2. MSE: 1.20 y RMSE: 1.10:
   * Como el MSE penaliza los errores grandes al elevarlos al cuadrado, nos dice que hay algo de variabilidad significativa en los errores.
   * El RMSE, que es más interpretable, dice que el error promedio (más realista) es de 1.10 unidades.
   * Esto refuerza lo que ya vimos con el MAE, pero mostrando que algunos errores podrían ser un poco más grandes.

3. R²: 0.40
   * Esto significa que tu modelo explica el 40% de la variabilidad del target 𝑦.
   * No es un mal comienzo, pero indica que hay bastante margen de mejora.
   * El modelo sí está captando cierta relación entre 𝑥 e 𝑦, pero no toda.

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Valores reales y predichos (pueden venir del modelo anterior)
y = np.array([3, 4, 2, 5, 6])
y_pred = np.array([2.8, 3.6, 4.4, 5.2, 6.0])  # ejemplo

# Cálculo de métricas
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y, y_pred)

print(f"MAE: {mae:.2f}")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R²: {r2:.2f}")
```

### 1.4 Regresión Polinómica

* Es una extensión de la regresión lineal que permite modelar relaciones no lineales elevando la variable de entrada 𝑥 a potencias mayores.
* Aunque se llama "polinómica", sigue siendo lineal en los coeficientes, lo que significa que se puede resolver con los mismos métodos matemáticos que la regresión lineal.
* ¿Qué hace PolynomialFeatures?
  - Transforma tu variable 𝑥 original en un conjunto de nuevas variables: 𝑥, 𝑥2, 𝑥3, etc.
  - Luego se entrena una regresión lineal común sobre estos nuevos datos.
* ¿Cómo saber si usar regresión polinómica?
  - Si los residuos de tu regresión lineal no se distribuyen aleatoriamente (forman curvas).
  - Si visualmente ves que una línea recta no ajusta bien.
  - Si el 𝑅2 mejora notablemente al usar un modelo no lineal.

* Ecuación del modelo (grado 2)
$$
y = \beta_0 + \beta_1 x + \beta_2 x^2 + \varepsilon
$$

* Ecuación del modelo (grado 3)
$$
y = \beta_0 + \beta_1 x + \beta_2 x^2 + \beta_3 x^3 + \varepsilon
$$

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Datos (no lineales)
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 6, 4, 10, 12])

# Transformar a polinomio de grado 2
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)

# Entrenar el modelo
model = LinearRegression()
model.fit(X_poly, y)

# Predicciones
y_pred = model.predict(X_poly)

# Mostrar coeficientes
print("Coeficientes:", model.coef_)
print("Intercepto:", model.intercept_)

# Gráfica
plt.scatter(X, y, color='blue', label='Datos reales')
plt.plot(X, y_pred, color='red', label='Regresión polinómica')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Regresión Polinómica (grado 2)')
plt.show()
```

# 2. Regresión

### Regresión basada en árboles

* Los modelos basados en árboles son muy potentes para problemas de regresión no lineales. ***No requieren que las relaciones entre variables*** sean lineales ni que las variables estén escaladas.
* ¿Qué veremos en este apartado?
  - Árboles de decisión para regresión.
  - Random Forest Regressor.
  - Gradient Boosting (XGBoost, LightGBM).

### Árboles de decisión para regresión

* Es un modelo que divide recursivamente el espacio de entrada en regiones, y predice el valor promedio de los datos en cada región.
* Si estás prediciendo el precio de una casa, el árbol puede hacer divisiones como:
  - ¿Superficie > 120 m²?
  - ¿Nº habitaciones > 3?
  - ¿Está en zona urbana?
* ***Cada hoja del árbol da una predicción*** (el promedio de los valores de esa hoja).
* Ventajas:
  - No requiere normalización.
  - Interpretable (se puede graficar).
  - Captura relaciones no lineales.
* Desventajas:
  - Sobreajuste si el árbol es muy profundo.
  - Menor precisión que otros métodos en algunos casos.

***Notas de Código***
* `DecisionTreeRegressor`: es el modelo de la clase `Tree`.
* ¿Cómo funciona make_regression?
  - `make_regression` es una función de `scikit-learn` que genera un conjunto de datos sintético `para problemas de regresión`.
  - n_samples=200: 200 puntos de datos, n_features=1: solo una variable independiente (fácil de visualizar), noise=15: agrega ruido gaussiano al target y para que no sea perfectamente lineal, random_state=42: semilla para reproducibilidad.
* ¿Por qué se pone un max_depth?
  - El hiperparámetro max_depth controla la profundidad máxima del árbol de decisión.
  - max_depth=1: árbol muy simple (1 división).
  - max_depth=3: árbol con hasta 3 niveles de decisión.
  - max_depth=None: el árbol crece hasta que cada hoja es pura (lo que puede causar sobreajuste).
  - Para evitar que el árbol memorice demasiado y generalice mal en datos nuevos (overfitting).
* ¿Por qué se calculó el MSE?
  - Es una forma estándar de evaluar qué tan bien el modelo está prediciendo. Cuanto menor el MSE, mejor se están aproximando las predicciones al valor real.
* ¿Por qué se ordenaron los datos antes de graficar?
  - Esto se hace solo para visualizar mejor la curva de predicción.
  - Los árboles de decisión generan predicciones que parecen escalones (piecewise constant).
  - Si graficamos sin ordenar, los puntos estarían desordenados y la curva se vería como un "garabato".
  -  Al ordenar X_test, nos aseguramos de que la línea de predicción se vea suave y clara, como una función que se puede seguir visualmente.
* ¿Ese MSE es "bueno" o "malo"?
  - Como estamos usando datos sintéticos con make_regression y noise=15, es esperado que el MSE no sea 0. El ruido hace que incluso un modelo perfecto no pueda predecir exactamente todos los valores.
  - Un MSE de 828.70 podría ser razonable para este nivel de ruido y con un árbol de profundidad limitada (max_depth=3). De hecho, si aumentaras la profundidad, el MSE en train bajaría, pero podrías caer en overfitting.
  - Mientras más max_depth, menor es el MSE.

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
import numpy as np

# Generamos datos de ejemplo
X, y = make_regression(n_samples=200, n_features=1, noise=15, random_state=42)

# División en train y test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Modelo
tree = DecisionTreeRegressor(max_depth=5, random_state=42)
tree.fit(X_train, y_train)

# Predicción
y_pred = tree.predict(X_test)

# Evaluación
mse = mean_squared_error(y_test, y_pred)
print(f"MSE del árbol de decisión: {mse:.2f}")

# Ordenamos para graficar bien
X_sorted = np.sort(X_test, axis=0)
y_sorted_pred = tree.predict(X_sorted)

plt.scatter(X_test, y_test, color='blue', label='Datos reales')
plt.plot(X_sorted, y_sorted_pred, color='red', label='Predicción árbol')
plt.legend()
plt.title("Árbol de decisión para regresión")
plt.show()
```

### Random Forest Regressor

* `Random Forest` es un conjunto de árboles de decisión entrenados sobre distintos subconjuntos del dataset. Luego, promedia sus predicciones para dar el resultado final
* Es un modelo más robusto y preciso que un solo árbol.
* Generaliza mejor (menos overfitting que un árbol solo).
* Funciona bien con muchos datos y variables.
* No necesita escalado.
* `from sklearn.ensemble import RandomForestRegressor` importación del modelo.
* `rf = RandomForestRegressor(n_estimators=100, max_depth=5, random_state=42)`: n_estimators=100: número de árboles, max_depth=5: controla la profundidad máxima de cada árbol, random_state=42: semilla reproducible.
* ¿Cuándo usar Random Forest?
  - Cuando un solo árbol no da buen resultado.
  - Cuando hay muchas variables y relaciones no lineales.
  - Cuando necesitas buen rendimiento sin mucho ajuste manual.

***NOTAS***
* ¿Random Forest usa la misma profundidad para todos los árboles?
  - Cuando defines max_depth=5, todos los árboles del Random Forest, tienen permitido crecer hasta una profundidad máxima de 5. Pero no necesariamente todos lo hacen. Algunos árboles pueden ser más pequeños si no lo necesitan (depende del subset de datos que reciben).
  - Es como decir: “todos pueden crecer hasta 5 pisos, pero no tienen que hacerlo”.
* ¿Cómo saber si no hay relaciones lineales en los datos?
  - Si haces un scatter plot de y vs cada X_i y ves curvas, escalones, picos, o formas raras, entonces la relación no es lineal.
  - Si el **modelo de regresión lineal tiene mucho error**, **pero** un **árbol o Random Forest mejora mucho**, es señal de que **la relación no es lineal**.
  - Probar un modelo lineal como referencia (baseline). Luego probar modelos no lineales y comparar métricas (como el MSE). Si mejoran significativamente, es que la relación no era lineal.
* ¿Cómo saber si un solo árbol no da buen resultado?
  - Entrenas un DecisionTreeRegressor y mides su MSE o R².
  - Luego entrenas un RandomForestRegressor y comparas.

* Modelos no lineales
  - Árbol de decisión
  - Random Forest
  - Gradient Boosting (XGB)
  - KNN
  - SVM (con kernel)
  - Redes neuronales

### Gradient Boosting para regresión.

* `Gradient Boosting` es una técnica que entrena modelos simples (como árboles poco profundos) uno tras otro, y cada nuevo modelo intenta corregir los errores del anterior.
* Es como si cada árbol aprendiera de los errores del árbol anterior.
* Se construye en secuencia, no en paralelo como Random Forest.
* Se enfoca en minimizar el error paso a paso.
* Es más preciso que Random Forest en muchos casos, pero más sensible a hiperparámetros.
* Algoritmos populares:
  - `xgboost`: Muy rápido, muy usado en competencias.
  - `lightgbm`:	Más rápido aún en grandes datasets.
  - `GradientBoostingRegressor`: de `sklearn` se usa por su implementación base, más fácil de usar.
* `from sklearn.ensemble import GradientBoostingRegressor`
* `gbr = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)`:
  - n_estimators: cuántos árboles se entrenan (más = más precisión, pero más lento).
  - learning_rate: cuánto corrige cada nuevo árbol.
  - max_depth: qué tan complejos son los árboles base.
* `Ventajas`: Preciso en muchos tipos de datos. Ideal para competiciones (Kaggle, etc.). Muy buen control de overfitting.
* `Desventajas`: Requiere ajustar hiperparámetros. Más lento en entrenamiento comparado con Random Forest. Menos interpretabilidad.
* ¿Cuándo usarlo?
  - Cuando necesitas alta precisión.
  - Cuando ya probaste árboles simples y Random Forest y quieres más.
  - Cuando tienes tiempo para ajustar bien hiperparámetros (con Grid Search, por ejemplo).

### 2.5 Regresión Lineal Múltiple

* Es una extensión de la regresión lineal simple, pero en lugar de tener una sola variable predictora (X), usamos varias al mismo tiempo.
* y=β0+ β1x1 + β2x2 + ⋯ + βnxn + ϵ.
* Cada 𝛽𝑖 representa el efecto individual de cada variable sobre el valor predicho 𝑦, manteniendo las otras constantes.
* Es la misma formula de la regresión pero agregando n variables.
* ¿Para qué sirve?
  - Para modelar fenómenos más complejos con múltiples factores (ej. predicción de precios de casas con tamaño, ubicación, número de habitaciones, etc.).
  - Para entender el impacto relativo de cada variable (coeficientes).
* `from sklearn.linear_model import LinearRegression`: se sigue usando el mismo modelo como con la diabetes.(eran varias variables)
* `Ventajas`: Sencillo y eficiente. Útil para explicar el impacto de cada variable. Rápido de entrenar.
* `Desventajas`: Asume relación lineal entre cada X y y. Puede sufrir de multicolinealidad (cuando las variables están correlacionadas entre sí). No maneja outliers o relaciones no lineales tan bien.
* ¿Cuándo usar regresión múltiple?
  - Cuando tienes varias variables numéricas y crees que hay una relación lineal.
  - Cuando buscas interpretabilidad más que máxima precisión.
  - Como modelo base antes de probar modelos más complejos.

### 2.6 Regularización: Ridge y Lasso

* ¿Qué es la regularización?
  - Es una técnica que se usa para evitar el sobreajuste (overfitting) en modelos de regresión, penalizando los coeficientes grandes.
  - Cuando tienes muchas variables o algunas muy correlacionadas, los coeficientes pueden crecer mucho para ajustar bien los datos, pero esto reduce la capacidad del modelo para generalizar.
* Ridge (L2)
  - Agrega una penalización proporcional al cuadrado de los coeficientes.
  - No elimina variables, pero reduce sus valores
  - Útil cuando todas las variables aportan algo, aunque poco.
* Lasso (L1)
  - Agrega una penalización proporcional al valor absoluto de los coeficientes.
  - Puede hacer que algunos coeficientes sean exactamente cero → selección de variables automática.
  - Útil si sospechas que algunas variables no aportan nada.
* `from sklearn.linear_model import Ridge, Lasso`: importación.
* `ridge = Ridge(alpha=1.0)` y `lasso = Lasso(alpha=0.1)`: alpha: controla la fuerza de la penalización (más alto = más fuerte).
* ¿Cuándo usar cada uno?
  - Rigde: Muchas variables pequeñas pero útiles.
  - Lasso: Algunas variables no son importantes.
  - Ambos: Quieres prevenir overfitting en general.

```python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np

# Crear un dataset con ruido y muchas variables
X, y = make_regression(n_samples=200, n_features=10, noise=20, random_state=42)

# Separar en train y test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Modelos
lr = LinearRegression()
ridge = Ridge(alpha=1.0)
lasso = Lasso(alpha=0.1)

# Entrenar
lr.fit(X_train, y_train)
ridge.fit(X_train, y_train)
lasso.fit(X_train, y_train)

# Predicciones
y_pred_lr = lr.predict(X_test)
y_pred_ridge = ridge.predict(X_test)
y_pred_lasso = lasso.predict(X_test)

# Métricas
mse_lr = mean_squared_error(y_test, y_pred_lr)
mse_ridge = mean_squared_error(y_test, y_pred_ridge)
mse_lasso = mean_squared_error(y_test, y_pred_lasso)

# Mostrar resultados
print("----- MSE -----")
print(f"Lineal: {mse_lr:.2f}")
print(f"Ridge:  {mse_ridge:.2f}")
print(f"Lasso:  {mse_lasso:.2f}")

print("\n----- Coeficientes -----")
print("Lineal:", np.round(lr.coef_, 2))
print("Ridge: ", np.round(ridge.coef_, 2))
print("Lasso: ", np.round(lasso.coef_, 2))
```

# 3. Modelos de Clasificación

* Es el proceso de predecir clases o categorías a partir de datos. A diferencia de regresión (que predice números continuos), aquí los resultados son discretos.
* Ejemplos:
  - ¿Es spam o no spam? (binario)
  - ¿Qué tipo de flor es? (multiclase)
  - ¿Tiene o no diabetes?

### 3.1 Regresión Logística

* Aunque suene a "regresión", es un modelo para clasificación binaria.
* Aprende a predecir la probabilidad de que una muestra pertenezca a la clase 1 (por ejemplo, "tiene diabetes").
* Usa la función sigmoide para convertir una predicción lineal en una probabilidad entre 0 y 1.
$$
\sigma(z) = \frac{1}{1 + e^{-z}}
$$
* Si σ(z)>0.5 → clase 1 ; σ(z)<0.5 → clase 0
* Métricas comunes en clasificación:
  - `Accuracy`: % de predicciones correctas
  - `Precision`: ¿Cuántos de los positivos predichos son correctos?
  - `Recall`: ¿Cuántos de los positivos reales fueron detectados?
  - `F1 Score`: Promedio balanceado de precision y recall
* ¿Cuándo usar Regresión Logística?
  - Cuando tienes dos clases y necesitas un modelo interpretable.
  - Para datos relativamente lineales y balanceados.
  - Como modelo base (baseline) en problemas de clasificación.

### 3.2 K-Nearest Neighbors (KNN)

* KNN es un modelo muy simple pero poderoso que clasifica una muestra según sus vecinos más cercanos.
* Idea central: Si quieres saber a qué clase pertenece un punto, mira qué clases tienen sus K vecinos más cercanos.
* ¿Cómo funciona?
  - Calcula la distancia entre el nuevo punto y todos los puntos del entrenamiento.
  - Toma los K vecinos más cercanos.
  - Asigna la clase más común entre ellos (voto mayoritario).
* Distancias más comunes:
  - Euclidiana (por defecto)
$$
d(x, x') = \sqrt{(x_1 - x_1')^2 + (x_2 - x_2')^2 + \cdots + (x_n - x_n')^2}
$$
  - Por eso, es muy sensible a la escala → se recomienda escalar los datos antes.
* `from sklearn.neighbors import KNeighborsClassifier`
* `Ventajas`: Muy fácil de entender y aplicar. No hace suposiciones sobre la distribución de los datos. Funciona bien si las clases están bien separadas.
* `Desventajas`: Muy sensible a escala y ruido. Rinde mal con datos de alta dimensión (muchas variables). Lento con grandes datasets (porque compara contra todos los puntos).
* ¿Cuándo usar KNN?
  - En datasets pequeños y bien estructurados.
  - Para problemas donde la relación entre puntos cercanos tiene sentido (ej. imágenes, geolocalización).
  - Como modelo base comparativo.

### 3.3 Árboles y Random Forest para Clasificación

* Árboles de Decisión (Decision Tree Classifier)
  - Un árbol de decisión clasifica los datos dividiéndolos por reglas basadas en sus características.
  - En cada nodo del árbol, se hace una pregunta:"¿feature_i ≤ valor?". Según la respuesta, seguimos por una rama u otra. El proceso se repite hasta llegar a una hoja donde se asigna una clase.
  - `from sklearn.tree import DecisionTreeClassifier`
  - `Ventajas`: Interpretable (puedes visualizar el árbol). No requiere escalar los datos. Maneja variables numéricas y categóricas.
  - `Desventajas`: Puede sobreajustar fácilmente si no se controla la profundidad o el número de muestras por nodo. Poca capacidad de generalización si el árbol es muy profundo.
* Random Forest Classifier
  - Es un conjunto de muchos árboles de decisión, entrenados sobre subconjuntos aleatorios del dataset y de sus variables.
  - La predicción final se toma por votación mayoritaria entre los árboles.
  - `Ventajas frente al árbol individual:` Más robusto y preciso. Menos overfitting. Funciona bien por defecto
  - `from sklearn.ensemble import RandomForestClassifier`

### 3.4 SVM – Máquinas de Vectores de Soporte

* SVM (Support Vector Machine) es un algoritmo de clasificación que busca encontrar el hiperplano que mejor separa las clases en el espacio de características.
* `Objetivo`: Maximizar el margen entre las clases. Es decir, encontrar la línea (o plano) que esté lo más lejos posible de los puntos más cercanos de cada clase.
* A estos puntos clave se les llama vectores de soporte.
* `Clasificación Lineal`: Cuando los datos son linealmente separables, SVM encuentra una línea recta (en 2D), o un plano (en 3D), que divide las clases.
* ` Clasificación No Lineal (Kernel Trick)`: Cuando los datos no pueden separarse linealmente, SVM usa funciones de transformación llamadas kernels para mapear los datos a un espacio donde sí se puedan separar.
* `from sklearn.svm import SVC`: La 'C' es por classifier.
* Ventajas
  - Muy efectivo en espacios de alta dimensión.
  - Funciona bien con márgenes claros de separación.
  - Puede ser muy preciso con buen ajuste.
* Desventajas
  - No es tan fácil de interpretar.
  - Puede ser lento en datasets grandes.
  - Sensible a la elección del kernel y los hiperparámetros (C, gamma, etc.).
* ¿Cuándo usar SVM?
  - Cuando las clases no son separables linealmente.
  - Cuando tienes pocos datos, pero bien separados.
  - Cuando necesitas un modelo fuerte y robusto sin necesidad de muchos árboles o redes.
