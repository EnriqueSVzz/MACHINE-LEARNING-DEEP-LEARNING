### La Maldición de la Dimensionalidad

### ¿Qué es?

* Es un conjunto de fenómenos que aparecen cuando los datos tienen muchas dimensiones (features/variables).
* A medida que aumenta la dimensión: El espacio crece exponencialmente.
  - Si tienes un cubo en 1D (una línea) → 10 puntos llenan el espacio.
  - En 2D (un cuadrado) → necesitas 10^2=100 puntos.
  - En 3D (un cubo) → 10^3=1000.
* En baja dimensión, podemos distinguir entre “cercano” y “lejano”.
* En alta dimensión, todas las distancias tienden a parecer iguales.
* Ejemplo: en 100 dimensiones, la diferencia entre el vecino más cercano y el más lejano puede ser insignificante.
* `Esto afecta a métodos como K-Means, KNN, DBSCAN, que dependen de distancias.`

* Sobreajuste (overfitting)
  - Con más dimensiones, los modelos pueden ajustarse demasiado a los datos de entrenamiento.
  - Se necesita mucho más número de muestras para generalizar bien.

* Dificultad en la visualización y la intuición
  - Es imposible imaginar más de 3 dimensiones.
  - La exploración y la interpretación de los datos se vuelve complicada.

### ¿Cómo se mitiga la maldición de la dimensionalidad?

1. Reducción de dimensionalidad
   - Técnicas como `PCA (Análisis de Componentes Principales)`, `t-SNE`, `UMAP`.
   - Sirven para comprimir datos en pocas dimensiones sin perder mucha información.

2. Selección de variables (Feature selection)

   - Eliminar variables irrelevantes o redundantes.
   - Usar algoritmos como Random Forest feature importance o mutual information.

3. Normalización/Escalado

   - Para que ninguna variable domine las distancias.

4. Modelos especializados para alta dimensión

   - Ejemplo: árboles de decisión, random forests, gradient boosting suelen manejar mejor la alta dimensión que K-Means.

### Generar datos aleatorios sintéticos de 3 variables

```python
import numpy as np

x = np.random.randint(10, size=10)
y = np.random.randint(10, size=10)
z = np.random.randint(10, size=10)
x, y, z
```

### La Maldición de la Dimensionalidad

```python
import matplotlib.pyplot as plt

#crear gráfica con tres subfiguras
fig = plt.figure(figsize=(18, 6))

#espacio unidimensional
ax = fig.add_subplot(1, 3, 1)
ax.set_title("1 Dimensión", fontsize=20)
ax.plot(x, [0]*10, linewidth=0, marker="*",
        color="purple", markersize=20)

#espacio bidimensional
ax = fig.add_subplot(1, 3, 2)
ax.set_title("2 Dimensión", fontsize=20)
ax.scatter(x, y, marker="*", c="purple", s=200)

#espacio tridimensional
ax = fig.add_subplot(1, 3, 3, projection="3d")
ax.set_title("3 Dimensión", fontsize=20)
ax.scatter(x, y, z, marker="*", c="purple", s=200)

plt.show()
```

### Exponencialmente

```python
fig = plt.figure(figsize=(18, 6))

# Espacio unidimensional
ax = fig.add_subplot(1, 3, 1)

ax.set_title("10^1 = 10", fontsize=35)

### Llenamos todos los valores.
ax.plot(range(0, 10), [0]*10, 
        marker="o", linewidth=0,
        color="skyblue", markersize=20)

ax.plot(x, [0]*10, linewidth=0,
        marker="*", color="purple",
        markersize=20)


# Espacio bidimensional
ax = fig.add_subplot(1, 3, 2)

ax.set_title("10^2 = 100", fontsize=35)

for i in range(10):
    ax.scatter(range(0, 10), [i]*10,
               s=200, c="skyblue", marker="*")
    
ax.scatter(x, y, marker="*", c="purple", s=200)


# Espacio tridimensional
ax = fig.add_subplot(1, 3, 3, projection="3d")

ax.set_title("10^3 = 1000", fontsize=35)

for i in range(10):
    for j in range(10):
         ax.scatter(range(0, 10), [i]*10, j, 
                    s=200, c="skyblue", marker="*",
                    alpha=0.25)

ax.scatter(x, y, z, marker="*", c="purple", s=200)

plt.show()
```
