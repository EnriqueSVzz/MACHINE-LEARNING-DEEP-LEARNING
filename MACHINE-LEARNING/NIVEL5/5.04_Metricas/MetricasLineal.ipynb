### <span style="color:red">Error Absoluto Máximo</span>

* Simplemente cual es la diferencia máxima entre el valor predicho con el valor real.
* Es decir se sacan todas las diferencias de todos los puntos para entrar la mayor diferencia.
* Como se saca una diferencia, es importante que sea absoluto para que no haya sesgos.
* El ***Error Absoluto Máximo (EAM)*** es una medida de precisión que indica el mayor error absoluto entre los valores predichos por un modelo y los valores reales. Es decir, mide cuál fue el peor error (en valor absoluto) que cometió el modelo.
* Indica qué tan mal puede fallar tu modelo en el peor caso observado.
$$
\text{EAM} = \max_{i} \left| y_i - \hat{y}_i \right|
$$

```python
from sklearn.metrics import max_error


y_verdadero = [1,2,3,4,5]
y_predicho = [1,2,3,4,-5]
max_error(y_verdadero,y_predicho)
```

### <span style="color:red">Error Absoluto Medio MAE</span>

* Viene del inglés Mean Absolute Error
* En lugar de sacar la diferencia más grande, este directamente promedia todas las diferencias.
* Es una métrica de evaluación usada para medir la precisión de un modelo de predicción. Indica, en promedio, cuánto se desvía la predicción del valor real, sin importar la dirección del error (positiva o negativa).
$$
\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} \left| y_i - \hat{y}_i \right|
$$
* MAE = 0 significa predicciones perfectas.
* Cuanto más alto sea el MAE, más lejos está tu modelo de los valores reales.
* A diferencia del RMSE (Root Mean Squared Error), el MAE no penaliza tanto los errores grandes, ya que no eleva al cuadrado los errores.

```python
from sklearn.metrics import mean_absolute_error

y_verdadero = [1,2,3,4,5]
y_predicho = [1,2,3,4,-5]
mean_absolute_error(y_verdadero,y_predicho)
```

### <span style="color:red">Error Cuadrático Medio MSE</span>

* El Error Cuadrático Medio (MSE) — Mean Squared Error — es una de las métricas más comunes para evaluar la calidad de un modelo de regresión.
* Mide el promedio de los errores al cuadrado.
* Penaliza más fuertemente los errores grandes que los pequeños.
* Siempre es no negativo.
* Es el promedio del RSS: $$\frac{RSS}{n}$$
* Su raíz cuadrada da el RMSE (Root Mean Squared Error):  $$\sqrt{MSE}$$
* ¿Cuándo usar MSE?
  - Cuando quieres penalizar errores grandes.
  - En regresión lineal, redes neuronales y muchos algoritmos que minimizan la función de pérdida cuadrática.
$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} \left( y_i - \hat{y}_i \right)^2
$$

```python
from sklearn.metrics import mean_squared_error

y_verdadero = [1,2,3,4,5]
y_predicho = [1,2,3,4,-5]
mean_squared_error(y_verdadero,y_predicho)*len(y_predicho)
```

### <span style="color:red">Suma de los Cuadrados de los Residuos (RSS)</span>

* La Suma de los Cuadrados de los Residuos (RSS, por sus siglas en inglés: Residual Sum of Squares) es una métrica fundamental en estadística y aprendizaje automático para evaluar qué tan bien se ajusta un modelo a los datos.
* La RSS mide la suma total de los errores al cuadrado entre los valores reales y los valores predichos por un modelo.
* Al elevar al cuadrado, penaliza más los errores grandes que los pequeños.
* Sirve para calcular métricas derivadas como el Error Cuadrático Medio (MSE), el RMSE, y el coeficiente de determinación 𝑅2
* Cuanto más pequeño es el RSS, mejor se ajusta el modelo a los datos.

$$
\text{RSS} = \sum_{i=1}^{n} \left( y_i - \hat{y}_i \right)^2
$$

```python
from sklearn.metrics import mean_squared_error

y_verdadero = [1,2,3,4,5]
y_predicho = [1,2,3,4,-5]
mean_squared_error(y_verdadero,y_predicho)*len(y_predicho)
```

### <span style="color:red">Raíz Cuadrada del Error Cuadrático Medio RMSE</span>

* La Raíz Cuadrada del Error Cuadrático Medio (RMSE) — Root Mean Squared Error — es una métrica que indica, en promedio, cuánto se desvían las predicciones del modelo respecto a los valores reales, expresada en las mismas unidades que los datos originales.
* Es simplemente la raíz cuadrada del MSE.
* Te dice, en promedio, cuánto se equivoca el modelo por observación.
* Tiene la misma escala que los datos originales, a diferencia del MSE (que está en unidades al cuadrado).
* Penaliza errores grandes más fuertemente que el MAE.
* ¿Cuándo usar RMSE?
  - Cuando los errores grandes deben tener mayor impacto en la evaluación.
  - En regresión lineal, series temporales y redes neuronales.

$$
\text{RMSE} = \sqrt{ \frac{1}{n} \sum_{i=1}^{n} \left( y_i - \hat{y}_i \right)^2 }
$$

```python
from sklearn.metrics import root_mean_squared_error

y_verdadero = [1,2,3,4,5]
y_predicho = [1,2,3,4,-5]
root_mean_squared_error(y_verdadero,y_predicho)
```

### <span style="color:red">R^2 Coeficiente de Determinación</span>

* Esta metrica no aplica para modelos no lineales, es decir cuadraticos etc.
* El 𝑅^2, o Coeficiente de Determinación, es una métrica estadística que mide qué tan bien un modelo de regresión explica la variabilidad de la variable dependiente.
* En otras palabras, te dice qué porcentaje de la variación en los datos reales está explicado por el modelo.
*  Interpretación de
  - R 2=1: el modelo predice perfectamente los datos.
  - R2=0: el modelo no explica nada; equivale a predecir la media.
  - R2<0: el modelo es peor que simplemente predecir el promedio.
* Un alto 𝑅2 no garantiza un buen modelo. Puede ser alto incluso con un modelo que no generaliza bien.
* No es adecuado para evaluar modelos no lineales complejos sin ajustarlo.
* Existe un R2 ajustable.


$$
R^2 = 1 - \frac{\text{RSS}}{\text{TSS}}
$$
$$
\text{TSS} = \sum_{i=1}^{n} \left( y_i - \bar{y} \right)^2
$$
$$
R^2 = 1 - \frac{
\sum_{i=1}^{n} \left( y_i - \hat{y}_i \right)^2
}{
\sum_{i=1}^{n} \left( y_i - \bar{y} \right)^2
}
$$

```python
from sklearn.metrics import r2_score

y_verdadero = [1,2,3,4,5]
y_predicho = [1,2,3,4,-5]
r2_score(y_verdadero,y_predicho)
```

### <span style="color:red">R^2 Coeficiente de Determinación (Ajustado)</span>

* El 𝑅2 ajustado es una versión modificada del coeficiente de determinación 𝑅2 que corrige el sesgo de aumentar automáticamente al agregar más variables al modelo. Esto es útil porque a veces al añadir variables innecesarias, el 𝑅2 tradicional mejora aunque el modelo no mejore realmente.
* n: número de observaciones.
* p: número de variables independientes (predictoras) en el modelo.

$$
R^2_{\text{ajustado}} = 1 - \left(1 - R^2\right) \frac{n - 1}{n - p - 1}
$$

```python
import numpy as np
from sklearn.metrics import r2_score

# Datos
y_verdadero = np.array([1, 2, 3, 4, 5])
y_predicho = np.array([1, 2, 3, 4, -5])

# R² clásico
r2 = r2_score(y_verdadero, y_predicho)

# R² ajustado
n = len(y_verdadero)      # número de observaciones
p = 1                     # número de variables predictoras (ajusta según tu modelo)

r2_ajustado = 1 - (1 - r2) * (n - 1) / (n - p - 1)

print("R2:", r2)
print("R2 ajustado:", r2_ajustado)
```
