# Unir y Separar Numpy Arrays

```python
import numpy as np

### Definimos los siguientes arrays
altura = np.array([1.74, 1.8, 1.78, 1.68, 1.78])
peso = np.array([81.4, 88.7, 87.3, 62.7, 81.6])
```

## Stack

* Podemos agrupar por `eje X (axis=0)` o `Y (axis=1)`.
* Para el stack por eje X, es como si se apilaran de arriba a abajo.
* Por otro lado, con eje y, es como si los pararas y los pones a lado de izquiera a derecha.
* El valor default siempre será `axis=0`.

```python
print("Agrupamos por Eje Horizontal: \n", np.stack((altura,peso), axis=0))
print("Agrupamos por Eje Vertical: \n",np.stack((altura,peso), axis=1))

### o simplemente concatenar
print("Concatenado: \n", np.concatenate((altura,peso)))
```

## Eje X

```python
### Usemos el siguiente stack horizontal
union = np.stack((altura,peso), axis=0)
print("Union: \n",union, "\nTipo: ", type(union))

### Separados, estonos va a separar en 2 varibles tipos array pero dentro de una lista
separados = np.split(union, 2)
print("Separados: \n",separados, "\nTipo: ", type(separados))
```

## Eje Y

```python
union = np.stack((altura,peso), axis=1)
print("Union: \n",union, "\nTipo: ", type(union))

### Como ya tiene 2 columnas, nos marca error, entonces es mejor separarlos por 5
separados = np.split(union, 5)
print("Separados: \n",separados, "\nTipo: ", type(separados))
```
