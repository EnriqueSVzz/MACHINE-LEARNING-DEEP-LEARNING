# Creación y Manipulación de DataFrames

## Creación de un DataFrame

````python
### Importamos nuestro pandas
import pandas as pd
````

## Inicializacion de un dataframe

* Imagenemos que tenemos los siguientes datos
* Nota:
  - Cada elemento de un data frame es una serie.
  - De esta manera solo lista los valores `paises = pd.DataFrame(datos)`.
  - Es decir no tiene contexto.

````python
nombre_paises = ["China", "India", "Estados Unidos", "Indonesia", "Paskitán", "Brasil",
                "Nigeria", "Bangladesh", "Rusia", "México"]

encabezado = ["poblacion", "porcentaje"]

datos = [[1439, 18.47], 
         [1380 ,17.70], 
         [331, 4.25], 
         [273, 3.51], 
         [220, 2.83],
         [212, 2.73], 
         [206, 2.64], 
         [164, 2.11], 
         [145, 1.87], 
         [128, 1.65]]

### Creamos el data frame de esta forma le pone más contexto
paises = pd.DataFrame(datos, index=nombre_paises, columns=encabezado)
paises
````

### Transformar un diccionario a DataFrame

````python
### Imaginemos que tenemos este diccionario

diccionario = {
    "China": [1439, 18.47],
    "India": [1380, 17.70],
    "Estados Unidos": [331, 4.25],
    "Indonesia": [273, 3.51],
    "Paskitán": [220, 2.83],
    "Brasil": [212, 2.73],
    "Nigeria": [206, 2.64],
    "Bangladesh": [164, 2.11],
    "Rusia": [145, 1.87],
    "México": [128, 1.65]
}

# Esto hará que este transpuesto, tenemos que establecer los indices
# ya que por default empezarán con 0 o 1.
paises = pd.DataFrame(diccionario, index=encabezado)
print("Normal: \n", paises)

### transponer el dataframe
paises = paises.transpose()
print("Transpuesto: \n", paises)
````

## Métodos de un DataFrame

````python
### Imprimir el tipo de datos de nuestro dataFrame
paises.dtypes
````

````python
### Accedemos la data
print("Valores: \n", paises.values)
print("Size: ", paises.size)
````

````python
print("Indices: \n", paises.index)
print("Columnas: ", paises.columns)
````

## Acceder a un elemento del DataFrame

* Podemos acceder a todos los valores de una columna, recordemos que también se imprimira su `index`.

````python
print(paises.poblacion)

### También esto funciona
print("\n", paises["poblacion"])
````

````python
### También podríamos imprimer una serie de columnas que nosotros queramos
paises[["poblacion", "porcentaje"]]
````

```python

paises.poblacion[0] # también marca error
```

````python
### Así accedemos a un elmento espefico, aunque esto ya no es correcto
paises["poblacion"][1]
````

````python
paises["poblacion"][0:3]
````

## Métodos básicos comunes de DataFrames

````python
### Así cambiamos el tipo de dato de una columna
paises["poblacion"] = paises["poblacion"].astype("int")
````

````python
### Así mostramos la descripción de la tabla
### Podemos ver como ya cambió el tipo de dato a int32
paises.info()
````

````python
### traernos los primeros 5 y los ultimos 5 por default, aunque también podemos especificar
paises.head(3), paises.tail(2)
````

* `paises.sort_values(by=["poblacion"], ascending=False, inplace=True)`, necesario para que se guarde.
* `ascending=True`: menor a mayor en ese orden.

````python
### Podemos ordenar, pero aquí tenemos que espeficar, creo que toma de menor a mayor por default
print("Paises: \n", paises)
ordenados = paises.sort_values(by=["poblacion"], ascending=True)
print("Paises Ordenados: \n", ordenados)
````

````python
# Si no nos interesara esto, podriamos ordenar por nuestro index
# es este caso los ordenara por orden alfabetico
paises.sort_index()
````

## Agregar y borrar Renglones y Columnas

````python
### tasa defertilidad por pais
tasa_fertalidad = [1.7, 2.2, 1.8, 2.3, 3.6, 1.7, 5.4, 2.1, 1.8, 2.1]

# Agregariamos estos valores como si se tratara de un diccionario
paises["tasa_fertalidad"] = tasa_fertalidad
paises
````

* Al hacer ` paises.pop("poblacion")` nos regresa la columna.
* `del` sí modifica el DataFrame directamente.
* No necesita `inplace=True` porque no es un método de pandas, es una palabra reservada de Python.
* Por ultimo `.drop()`, este borra como tazas y columnas, `axis=1` es una columna, `axis=0` es un renglon.
* Por si solo, no la modifica, tenemos que activar que la modifica
* ***¿Cuándo usar cada uno?***
  - `del` → cuando solo quieres borrar y no necesitas la columna.
  - `pop()` → cuando quieres borrar y además guardar la columna.
  - `drop()` → cuando quieres más control (varias columnas, filas, etc.).

````python
### Así podemos eliminar una columna, y te muestra la columna que borro
paises.pop("tasa_fertalidad")
````

````python
### hace un borrado pero no muestra lo que borro como el anterior
del paises["tasa_fertalidad"]
````

````python
paises
````

````python
paises.drop("tasa_fertalidad", axis=1, inplace=True)
````

````python
### Agregar un valor, como los data frames están hechos de series, tenemos que dar de alta una
renglon = pd.Series(name="Japón", data=[126,1.62], index=["poblacion", "porcentaje"])
print("Nueva data :\n", renglon)
#paises.append(renglon) # esto ya esta desactualizado

### Aquí convertimos la serie a dataframe de una fila, y lo concatemos con paises
### Hay que transponerla para que tenga forma horizontal como una fila de excel
paises = pd.concat([paises, renglon.to_frame().T])
paises
````

````python
### Ahora hagamos que borre varios datos por fila
paises.drop(["Japón", "Bangladesh"], axis = 0, inplace=True)
````

## Funciones de Estadísitica y de agregación básica

````python
### igual que las series podemos pedir un pequeño resumen
print("Describe: \n", paises.describe())
print("\n Min: \n", paises.min())
````

````python
### podemos pedir acumulativos, se va sumando el porcentaje y la poblacion
paises.cumsum()
````
