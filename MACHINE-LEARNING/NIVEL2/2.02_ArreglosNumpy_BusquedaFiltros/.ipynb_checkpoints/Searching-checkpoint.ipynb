# Busquedas en Arreglos

```python
import numpy as np
```

## Busquedas

```python
### definimos nuestro array
arreglo = np.array([11,12,13,14,15])

### imprimimos nuestros valores mayor a 13
print(np.where(arreglo >= 13))

### También podemos establecer directamente un vector booleano, la relación tiene que ser 1:1
filtro = [True, False, False, True, True]
print(arreglo[filtro])
```

## EJERCICIO

```python
### Imaginemos que tenemos estos datos.
altura = np.array([1.74, 1.8, 1.78, 1.68, 1.78])
peso = np.array([81.4, 88.7, 87.3, 62.7, 81.6])
pais = np.array(["México", "USA", "Canadá", "China", "Perú"])

### Aplicamos el siguiente filtro, todos los paises cuyos, pesos pasen de 90
print("Paises con Peso >85: ", pais[peso>85], " Total: ", pais[peso>85].size)
print("Paises con Altura >1.75: ", pais[altura>1.75], " Total: ", pais[altura>1.75].size)



### Multiplicaciónes vectoriales, cada quien con el suyo
print("Resultado Booleano: ", np.array([True, False]) & np.array([False, False]))

### dado lo anterior, ahora también podríamos hacer esto
### debemos poner los parentesis, sino marca error en las condiciones
print(pais[(peso>85) & (altura>1.75)], pais[(peso>85) & (altura>1.75)].size)
print(pais[(peso>85) | (altura>1.75)], pais[(peso>85) | (altura>1.75)].size)
```
